Теперь построим график заданной кривой:

Теперь найдем асимптоты (при наличии) и также добавим их на график:

**Алгоритм построения ассимптот.**

Ассимптота - прямая, обладающая тем свойством, что расстояние от точки: $(x, f(x))$ графика функции до этой прямой стремится к нулю при неограниченном удалении точки графика от начала координат. Задается уравнением вида:
$$ y = kx +  b $$
У функции может не быть ассимптот, а может и быть бесконечно много, тут так сразу не угадаешь. Благо, жизнь нам предлагает алгоритм по нахождению ассимптот функции, при этом этот же алгоритм позволяет доказать, что ассимптот не существует.

Ассимптота существует, если оба предела: $$\lim_{x\to \infty} \frac{f(x)}{x}=k$$  и  $$\lim_{x\to \infty} {f(x)-kx}=b$$ существуют и не равны бесконечности.

Но жестокая судьба так и шепчет, что не все так просто. И дейтсвительно, ведь функция, заданная выше, задана неявно, а значит, для нахождения ее ассимптот нужны более общие правила.

Имеем что-то вроде: $$F(x,y)=0 $$  Пусть асимптота описывается уравнением $y=kx+b$, тогда, подставляя это выражение для $y$ в уравнение кривой, получаем алгебраическое уравнение относительно одной переменной $x$:
$${A_0}{x^n} + {A_1}{x^{n - 1}} +  \ldots  + {A_{n - 1}}x + {A_n} = 0$$

где коэффициенты $A_i$зависят от параметров асимптоты $k$ и $b$ (причем коэффициент $A_0$ зависит лишь от $k$). Значения $k$ и $b$ определяются из условия:
$$\left\{ \begin{array}{l}
{A_0}\left( k \right) = 0\\
{A_1}\left( {k,b} \right) = 0
\end{array} \right..$$

Для нахождения вертикальной асимптоты нужно подставить ее уравнение $x=a$ в уравнение кривой и преобразовать последнее к виду:

$${B_0}{y^n} + {B_1}{y^{n - 1}} +  \ldots  + {B_{n - 1}}y + {B_n} = 0$$

Необходимым условием существования вертикальной асимптоты является отсутствие в последнем уравнении старшего члена ${B_0}{y^n}$ Значение параметра $a$ определяется из условия:
$${B_1}\left( a \right) = 0$$

Приведенные формулы для асимптот неявно заданных кривых справедливы, если кривая не имеет особых точек на бесконечности.

Подставим $ y = kx+b$ в уравнение и приведем подобные:

```python
# Создаём новые переменные $k$ и $b$:
var('k b')
# Выполняем подстановку
curve2d_a = curve2d(y = k * x + b).expand()
# Выводим результат:
curve2d_a
```

Далее возьмем коэфициенты при двух старших степенях х.

```python
# Получаем коэффициенты при $x$ в порядке от $x^0$ до $x^2$:
coefficents = curve2d_a.coefficients(x, sparse=False)
# Выбираем старшие два:
coefficents = coefficents[-2:]
# Выводим результат:
coefficents
```

И, наконец, решим систему из полученных уравнений, найдя k и b:

```python
# Решаем систему, получаем список содержащих решение словарей:
asymptotes = solve(coefficents, k, b, solution_dict=True)
# Выводим результат:
asymptotes
```

```python
# Сразу переводим полученные коэффициенты $k$ и $b$ в асимптоты:
for i in range(len(asymptotes)):
    asymptotes[i] = asymptotes[i][k] * x + asymptotes[i][b]
# Выводим результат:
asymptotes
```

Теперь остается добавить на график обе ассимптоты:

```python
# Задаём цвета для ассимптот:
asymptote_colours = ["#61bfb6", "#b661bf"]
# Добавляем каждую на график:
for i in range(len(asymptotes)):
    plot2d += plot(
            asymptotes[i], (-5,5),
            xmin=-5, xmax=5, ymin=-5, ymax=5,
            color=asymptote_colours[i], linestyle = '--')
# Выводим график:
plot2d
```

1. Теория по построению асимптот неявно заданной функции
2. Задание на нахождение производной неявно заданной функции
3. Построить график производной функции, которая задает кривую 2-го порядка (синим цветом выделить промежутки убывания функции, красным - возрастания).
4. Заставить найти точку пересечения асимптот

```python
# Ищем производную неявно заданной функции:
curve2d_derivative = -curve2d.diff(x) / curve2d.diff(y)
# Выводим результат:
curve2d_derivative
```

```python
# Создаём график производной, где синим цветом
# обозначено убывание, а красным -- возрастание исходной функции
# (не её производной, а именно исходной функции -- я так понял условие):
plot2d_coloured = implicit_plot(
        curve2d_derivative, (x,-5,5), (y,-6,-0),
        xmin=-5, xmax=5, ymin=-5, ymax=5, plot_points=300,
        color="#61bfb6")
plot2d_coloured += implicit_plot(
        curve2d_derivative, (x,-5,5), (y,0,6),
        xmin=-5, xmax=5, ymin=-5, ymax=5, plot_points=300,
        color="#bf616a", axes=True)
# Выводим график:
plot2d_coloured
```

```python
# Создаём однотонный график производной для проверки корректности:
plot2d_derivative = implicit_plot(
        curve2d_derivative, (x,-5,5), (y,-6,6),
        xmin=-5, xmax=5, ymin=-5, ymax=5, plot_points=600,
        color="#bf616a")
# Выводим график:
plot2d_derivative
```

Осталось только найти точку пересечения ассимптот:

```python
# Приравниваем асимптоты, решаем полученное уравнение:
asymptote_intersections = solve(asymptotes[0] == asymptotes[1], x, solution_dict=True)
# Так как из геометрических соображений пересечение только одно,
# избавляемся от списка из одного элемента:
asymptote_intersection = asymptote_intersections[0]
# Находим $y$ исходя из одной из ассимптот:
asymptote_intersection[y] = asymptotes[0].subs(x=asymptote_intersection[x])
# Выводим результат:
asymptote_intersection
```

```python
# Добавляем на график, чтобы убедиться в корректности:
plot2d += point(
        (asymptote_intersection[x], asymptote_intersection[y]),
        color="#000", size=40)
# Выводим график:
plot2d
```

Для удобной работы с уравнением кривой второго порядка, само уравнения надо представить в виде суммы некоей квадратичной формы и линейного функционала. В данном случае часть уравнения, являющаяся КФ имеет вид:
                           $$ y^2−\frac{x^2}{10}−2\cdot x\cdot y $$
# Выводим результат:
Линейная часть выглядит так:
 $$ −x−y $$
Введем теперь вектор $v$:
$$ v = v_e = (x, y)^T $$
Тогда если обозначить матрицу квадратичной формы за $Q$, а матрицу линейного функционала за $L$, то наше уравнение примет следующий вид:
$$ v^T * Q * v + L*v = 0$$

Ваша текущая задача заключается в том, чтобы составить матрицы Q и L:
(Замечание: в SageMath существует объект QuadraticForm, который позволяет работать не с абстрактными матрицами, а с матрицами именно квадратичной формы. Решение, реализованное не по шаблону, с использованием QuadraticForm также будет приниматься, но рассматриваться в ноутбуке не будет).

```python
# Задаём вектор $L$:
L2d = vector( [-1, -1] )
# Выводим результат:
L2d
```

```python
# Задаём матрицу $Q$:
Q2d = matrix( [ [-1/10, -1], [-1, 1] ] )
# Выводим результат:
Q2d
```

```python
# Задаём вектор $v$:
v2d = vector( [x, y] )
# Выводим результат:
v2d
```

```python
# Выполняем проверку:
print((v2d * Q2d * v2d + L2d * v2d).expand())
curve2d
# Совпадение очевидно.
```

Теперь нам нужно диагонализировать матрицу КФ: для этого нам нужно перейти из стандартного базиса к базису из собственных векторов. Следовательно, следует найти собственные числа и векторы матрицы $Q$:

```python
# Находим собственные числа и вектора:
eigenvalues_and_more = Q2d.right_eigenvectors()
# Выводим результат:
eigenvalues_and_more
```

```python
# Находим собственные числа (с учётом кратности):
eigenvalues = sum([[t[0]] * t[2] for t in eigenvalues_and_more], [])
# Выводим результат:
eigenvalues
```

```python
# Находим собственные вектора:
eigenvectors = sum([t[1] for t in eigenvalues_and_more], [])
# Выводим результат:
eigenvectors
```

Получив собственные векторы матрицы можно задать матрицу перехода от стандартного базиса в $R^2$ к базису из собственных векторов. Запишите матрицу перехода, составив ее из собственных векторов и при необходимости избавившись от дробей в векторах (вектора следует нормировать):

```python
# Нормируем собственные вектора:
eigenvectors = [vector / vector.norm(2) for vector in eigenvectors]
# Выводим результат:
eigenvectors
```

```python
# Находим матрицу перехода:
transformation_matrix = matrix(eigenvectors).T
# Выводим результат:
transformation_matrix
```

Теперь проверьте равенство норм обоих собственных векторов (обратите внимение, здесь используется стандартная евклидова, т.е. вторая, норма):

```python
if eigenvectors[0].norm(2) == eigenvectors[1].norm(2):
    print("Верно,", eigenvectors[0].norm(2), '==', eigenvectors[1].norm(2))
else:
    print("Неверно,", eigenvectors[0].norm(2), '!=', eigenvectors[1].norm(2))
if transformation_matrix[0].norm(2) == transformation_matrix[1].norm(2):
    print("Верно,", transformation_matrix[0].norm(2), '==', transformation_matrix[1].norm(2))
else:
    print("Неверно,", transformation_matrix[0].norm(2), '!=', transformation_matrix[1].norm(2))
```

Перепишем теперь матрицу $Q$ в новом базисе:

```python
# Производим преобразование:
Q2d_transformed = transformation_matrix.inverse() * Q2d * transformation_matrix
# Выводим результат:
Q2d_transformed
```

Теперь подставим полученные переменные в наше исходное уравнение:

```python
# Производим преобразование:
L2d_transformed = transformation_matrix * L2d
# Выводим результат:
L2d_transformed
```

```python
# Собираем неявную функцию:
curve2d_transformed = (v2d * Q2d_transformed * v2d + L2d_transformed * v2d).expand()
# Выводим результат:
curve2d_transformed
```

Для проверки решения снова построим график:

```python
# Создаём график новой неявной функции:
plot2d_transformed = implicit_plot(
        curve2d_transformed, (x,-5,5), (y,-6,6),
        xmin=-5, xmax=5, ymin=-5, ymax=5,
        color="#bf616a", axes=True)
# Выводим график:
plot2d_transformed
```

```python
# Убираем оси (они больше не требуются)
plot2d_transformed = implicit_plot(
        curve2d_transformed, (x,-5,5), (y,-6,6),
        xmin=-5, xmax=5, ymin=-5, ymax=5, 
        color="#bf616a")
```

Если все было сделано верно, то мы должны были добиться поворота графика таким образом, чтобы он стал параллелен одной из осей координат. Остались только последние штрихи: отцентрировать график.

```python
# Выполняем подстановку
curve2d_transformed_a = curve2d_transformed(y = k * x + b).expand()
# Выводим результат:
curve2d_transformed_a
```

```python
# Получаем коэффициенты при $x$ в порядке от $x^0$ до $x^2$:
coefficents_transformed = curve2d_transformed_a.coefficients(x, sparse=False)
# Выбираем старшие два:
coefficents_transformed = coefficents_transformed[-2:]
# Выводим результат:
coefficents_transformed
```

```python
# Решаем систему, получаем список содержащих решение словарей:
asymptotes_transformed = solve(coefficents_transformed, k, b, solution_dict=True)
# Выводим результат:
asymptotes_transformed
```

```python
# Сразу переводим полученные коэффициенты $k$ и $b$ в асимптоты:
for i in range(len(asymptotes_transformed)):
    asymptotes_transformed[i] = asymptotes_transformed[i][k] * x + asymptotes_transformed[i][b]
# Выводим результат:
asymptotes_transformed
```

```python
# Добавляем каждую на график, чтобы убедиться в корректности:
for i in range(len(asymptotes_transformed)):
    plot2d_transformed += plot(
            asymptotes_transformed[i], (-5,5),
            xmin=-5, xmax=5, ymin=-5, ymax=5,
            color=asymptote_colours[i], linestyle = '--')
# Выводим график:
plot2d_transformed
```

```python
# Приравниваем асимптоты, решаем полученное уравнение:
asymptote_intersections_transformed = solve(asymptotes_transformed[0] == asymptotes_transformed[1], x, solution_dict=True)
# Так как из геометрических соображений пересечение только одно,
# избавляемся от списка из одного элемента:
asymptote_intersection_transformed = asymptote_intersections_transformed[0]
# Находим $y$ исходя из одной из ассимптот:
asymptote_intersection_transformed[y] = asymptotes_transformed[1].subs(x=asymptote_intersection_transformed[x])
# Выводим результат:
asymptote_intersection_transformed
```

```python
# Добавляем на график, чтобы убедиться в корректности:
plot2d_transformed += point(
        (asymptote_intersection_transformed[x], asymptote_intersection_transformed[y]),
        color="#000", size=40)
# Выводим график:
plot2d_transformed
```

Теперь путем параллельного переноса отцентрируем график гиперболы:

**Замечание.** Сдвигать надо на значение х и у, полученные полученные в предыдущем шаге(точки пересечения ассимптот функции после поворота)

```python
# Центрируем неявную функцию:
curve2d_centered = curve2d_transformed(x = x + asymptote_intersection_transformed[x], y = y + asymptote_intersection_transformed[y]).expand()
# Выводим результат:
curve2d_centered
```

```python
# Создаём график новой неявной функции:
plot2d_centered = implicit_plot(
        curve2d_centered, (x,-5,5), (y,-6,6), 
        xmin=-5, xmax=5, ymin=-5, ymax=5,
        color="#bf616a", axes=True)
# Выводим график:
plot2d_centered
```

Поздравляем! С помощью параллельного переноса график гиперболы был отцентрирован, а первая часть проекта завершена.

```python
# Выполняем проверку:
print((v3d * Q3d * v3d + L3d * v3d).expand())
curve3d
# Совпадение очевидно.
```

Теперь нам нужно диагонализировать матрицу КФ: для этого нам нужно перейти из стандартного базиса к базису из собственных векторов. Следовательно, следует найти собственные числа и векторы матрицы $Q$:

```python
# Находим собственные числа и вектора:
eigenvalues_and_more_3d = Q3d.right_eigenvectors()
# Выводим результат:
eigenvalues_and_more_3d
```

```python
# Находим собственные числа (с учётом кратности):
eigenvalues_3d = sum([[t[0]] * t[2] for t in eigenvalues_and_more_3d], [])
# Выводим результат:
eigenvalues_3d
```

```python
# Находим собственные вектора:
eigenvectors_3d = sum([t[1] for t in eigenvalues_and_more_3d], [])
# Выводим результат:
eigenvectors_3d
```

Получив собственные векторы матрицы можно задать матрицу перехода от стандартного базиса в $R^2$ к базису из собственных векторов. Запишите матрицу перехода, составив ее из собственных векторов и при необходимости избавившись от дробей в векторах (вектора следует нормировать):

```python
# Нормируем собственные вектора:
eigenvectors_3d = [vector / vector.norm(2) for vector in eigenvectors_3d]
# Выводим результат:
eigenvectors_3d
```

```python
# Находим матрицу перехода:
transformation_matrix_3d = matrix(eigenvectors_3d).T
# Выводим результат:
transformation_matrix_3d
```

Теперь проверьте равенство норм обоих собственных векторов (обратите внимение, здесь используется стандартная евклидова, т.е. вторая, норма):

```python
if eigenvectors_3d[0].norm(2) == eigenvectors_3d[1].norm(2):
    print("Верно,", eigenvectors_3d[0].norm(2), '==', eigenvectors_3d[1].norm(2))
else:
    print("Неверно,", eigenvectors_3d[0].norm(2), '!=', eigenvectors_3d[1].norm(2))
if eigenvectors_3d[0].norm(2) == eigenvectors_3d[2].norm(2):
    print("Верно,", eigenvectors_3d[0].norm(2), '==', eigenvectors_3d[2].norm(2))
else:
    print("Неверно,", eigenvectors_3d[0].norm(2), '!=', eigenvectors_3d[2].norm(2))
if eigenvectors_3d[1].norm(2) == eigenvectors_3d[2].norm(2):
    print("Верно,", eigenvectors_3d[1].norm(2), '==', eigenvectors_3d[2].norm(2))
else:
    print("Неверно,", eigenvectors_3d[1].norm(2), '!=', eigenvectors_3d[2].norm(2))
```

Перепишем теперь матрицу $Q$ в новом базисе:

```python
# Производим преобразование:
Q3d_transformed = transformation_matrix_3d.inverse() * Q3d * transformation_matrix_3d
# Выводим результат:
Q3d_transformed
```

Теперь подставим полученные переменные в наше исходное уравнение:

```python
# Производим преобразование:
L3d_transformed = transformation_matrix_3d * L3d
# Выводим результат:
L3d_transformed
```

```python
# Собираем неявную функцию:
curve3d_transformed = (v3d * Q3d_transformed * v3d + L3d_transformed * v3d).expand()
# Выводим результат:
curve3d_transformed
```

Для проверки решения снова построим график:

```python
# Создаём график новой неявной функции:
plot3d_transformed = implicit_plot3d(
        curve3d_transformed, (x,-10,10), (y,-10,10), (z,-10,10),
        xmin=-10, xmax=10, ymin=-10, ymax=10, zmin=-10, zmax=10,
        viewer='tachyon', adaptive = 'true', axes=True,
        color="#bf616a", opacity='0.85')
# Выводим график:
plot3d_transformed
# Полученный график выровнялся относительно осей.
```
