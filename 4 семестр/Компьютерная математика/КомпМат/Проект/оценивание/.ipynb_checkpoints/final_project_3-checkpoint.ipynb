## Кривая:
Задана следующая квадратичная форма:
    $$ y^2-x^2-xy+7x=0 $$

```python
#уравнение к.ф. в символьном виде
var('x y z k b a')
f(x,y) = y^2 - x^2 - x*y + 7*x
```

Теперь построим график заданной кривой:

```python
implicit_plot(f, (-50,50), (-50,50), axes_labels=[x,y], color='red')
```

Теперь найдем асимптоты (при наличии) и также добавим их на график:

**Алгоритм построения асимптот.**

Асимптота - прямая, обладающая тем свойством, что расстояние от точки: $(x, f(x))$ графика функции до этой прямой стремится к нулю при неограниченном удалении точки графика от начала координат. Задается уравнением вида:
$$ y = kx +  b $$
У функции может не быть асимптот, а может и быть бесконечно много, тут так сразу не угадаешь. Благо, жизнь нам предлагает алгоритм по нахождению асимптот функции, при этом этот же алгоритм позволяет доказать, что асимптот не существует.

Асимптота существует, если оба предела: $$\lim_{x\to \infty} \frac{f(x)}{x}=k$$  и  $$\lim_{x\to \infty} {f(x)-kx}=b$$ существуют и не равны бесконечности.

Но жестокая судьба так и шепчет, что не все так просто. И дейтсвительно, ведь функция, заданная выше, задана неявно, а значит, для нахождения ее асимптот нужны более общие правила.

Имеем что-то вроде: $$F(x,y)=0 $$  Пусть асимптота описывается уравнением $y=kx+b$, тогда, подставляя это выражение для $y$ в уравнение кривой, получаем алгебраическое уравнение относительно одной переменной $x$:
$${A_0}{x^n} + {A_1}{x^{n - 1}} +  \ldots  + {A_{n - 1}}x + {A_n} = 0$$

где коэффициенты $A_i$зависят от параметров асимптоты $k$ и $b$ (причем коэффициент $A_0$ зависит лишь от $k$). Значения $k$ и $b$ определяются из условия:
$$\left\{ \begin{array}{l}
{A_0}\left( k \right) = 0\\
{A_1}\left( {k,b} \right) = 0
\end{array} \right..$$

Для нахождения вертикальной асимптоты нужно подставить ее уравнение $x=a$ в уравнение кривой и преобразовать последнее к виду:

$${B_0}{y^n} + {B_1}{y^{n - 1}} +  \ldots  + {B_{n - 1}}y + {B_n} = 0$$

Необходимым условием существования вертикальной асимптоты является отсутствие в последнем уравнении старшего члена ${B_0}{y^n}$ Значение параметра $a$ определяется из условия:
$${B_1}\left( a \right) = 0$$

Приведенные формулы для асимптот неявно заданных кривых справедливы, если кривая не имеет особых точек на бесконечности.

Подставим $ y = kx+b$ в уравнение и приведем подобные:

```python
from sage.calculus.calculus import at

def change_of_variables(expr, lst): #функция предназначена для замены переменных в уравнении
    for elem in lst:
        expr = at(expr, lst)
    return expr

new_f1 = change_of_variables(f, [y == k*x + b]) #замена переменных в исходном уравнении
print("Уравнение после замены y = k*x + b:\t" + str(new_f1))
```

Далее возьмем коэфициенты при двух старших степенях х.

```python
A = new_f1.full_simplify() #упрощённый вид алгебраического уравнения для наглядности
print("Упрощенный вид уравнения после замены: " + str(A) + "\n")
A0 = new_f1.full_simplify().coefficient(x, 2) #коэффицент А0 при старшей степени x
A1 = new_f1.full_simplify().coefficient(x, 1) #коэффицент А1 при степени N-1 в x
print("Коэффициент A0 = " + str(A0) + "\n")
print("Коэффициент A1 = " + str(A1))
```

Решим систему из полученных уравнений, найдя k и b:

```python
new_k = A0.solve(k)
new_b = []
for i in new_k:
    urv = change_of_variables(A1, [i])
    new_b.append(urv.solve(b))
print(str(new_k[0]) + " <-- k1\n")
print(str(new_k[1]) + " <-- k2\n")
print(str(new_b[0][0]) + " <-- b1\n")
print(str(new_b[1][0]) + " <-- b2")
```

Теперь остается добавить на график обе асимптоты:

```python
#запишем две асимптоты вида y1 = k1*x + b1 и y2 = k2*x + b2
asymptote = k*x + b
asympt = []
for i in range(len(new_k)):
    asympt.append(change_of_variables(asymptote, [new_k[i], new_b[i][0]]))
function = implicit_plot(f, (-15,15), (-15,15), axes_labels=[x,y], color = 'red')
for i in asympt:
    function += plot(i, (-15, 15), color = 'blue')
show(function)
```

Осталось только найти точку пересечения асимптот:

```python
x0 = solve([asympt[0] == asympt[1]], x)[0] #координата x
y0 = change_of_variables(asympt[0], [x0]) #координата y
print(x0)
print("y == (" + str(y0) + ")")
```

```python
p = point((14/5,7/5), color = "black", size = 50) #точка пересечения асим 
show(function+p)
```

Для удобной работы с уравнением кривой второго порядка, само уравнения надо представить в виде суммы некоей квадратичной формы и линейного функционала. В данном случае часть уравнения, являющаяся КФ имеет вид:
                           $$ y^2 - x^2 - xy $$
Линейная часть выглядит так:
 $$ 7x $$
Введем теперь вектор $v$:
$$ v = v_e = (x, y)^T $$
Тогда если обозначить матрицу квадратичной формы за $Q$, а матрицу линейного функционала за $L$, то наше уравнение примет следующий вид:
$$ v^T * Q * v + L*v = 0$$

Текущая задача заключается в том, чтобы составить матрицы Q и L:

```python
v = matrix([x,y])
Q = matrix([[-1,-1/2],[-1/2,1]])
L = matrix([7, 0])
```

Теперь нам нужно диагонализировать матрицу КФ: для этого нам нужно перейти из стандартного базиса к базису из собственных векторов. Следовательно, следует найти собственные числа и векторы матрицы $Q$:

```python
vec2, vec1 = Q.right_eigenvectors() #получаем тройки значений и записываем их в переменные
vec1 = vector(vec1[1][0]) #получаем первый собственный вектор
vec2 = vector(vec2[1][0]) #получаем второй собственный вектор
print("первый вектор: " + str(vec1) + "\n")
print("второй вектор: " + str(vec2))
```

Получив собственные векторы матрицы можно задать матрицу перехода от стандартного базиса в $R^2$ к базису из собственных векторов. Запишите матрицу перехода, составив ее из собственных векторов и при необходимости избавившись от дробей в векторах (вектора следует нормировать):

```python
new_matrix = matrix([vec1, vec2])
print(new_matrix)
```

Теперь проверьте равенство норм обоих собственных векторов (обратите внимение, здесь используется стандартная евклидова, т.е. вторая, норма):

```python
if vec1.normalized().norm(2) == vec2.normalized().norm(2):
    print("Равенство норм верно")
else:
    print("Равенство норм не верно")
```

Перепишем теперь матрицу $Q$ в новом базисе:

```python
new_Q = new_matrix*Q*new_matrix^(-1)
print(new_Q)
```

Теперь подставим полученные переменные в наше исходное уравнение:

```python
equation = (v*new_Q*v.transpose()+L*v.transpose())[0][0]
print("Новое полученное уравнение: " + str(equation))
```

Для проверки решения снова построим график:
Если все было сделано верно, то мы должны были добиться поворота графика таким образом, чтобы он стал параллелен одной из осей координат. Остались только последние штрихи: отцентрировать график.

```python
turn = implicit_plot(equation, (-30,30), (-30,30), axes=True, axes_labels=[x,y], color="red")
show(turn) #график параллелен оси Oy
```

Теперь путем параллельного переноса отцентрируем график гиперболы:

Найдём коэффиценты ассимптоты:

```python
asympt_eq = change_of_variables(equation, [y == k*x + b])
coeffs = asympt_eq.coefficients(x, sparse=False)
coeff1 = coefficents[1]
coeff2 = coefficents[2]
print("первый коэффицент: " + str(coeff1))
print("второй коэффицент: " + str(coeff2))
coeff1 = -2.236067977499790*b*k + 7 #данная запись идентична полученным коеффицентам, но без знака '?'.
coeff2 = -1.118033988749895*k^2 + 1.118033988749895 #правильность вводимых коеф. можно посмотреть в выводе print().
```

Решим систему уравнений, чтобы найти k и b:

```python
sol = solve([coeff1, coeff2], k, b) #решаем систему относительно переменных k,b
equation_asympt = k*x + b
asympt1 = at(equation_asympt, sol[0])
asympt2 = at(equation_asympt, sol[1])
print("y1 = " + str(asympt1))
print("y2 = " + str(asympt2))
```

```python
x_center = solve(asympt1 == asympt2, x, solution_dict=True)[0][x] #находим x
print("x = " + str(x_center))
y_center = 0
y_center1 = asympt1.substitute(x = x_center) #подставляем в уравнение асимптот
y_center2 = asympt2.substitute(x = x_center)
if (y_center1 == y_center2):
    y_center = y_center1 #находим y
print("y = " + str(y_center))
```

```python
implicit_plot(equation(x=x+x_center,y=y+y_center), (-20,20), (-20,20), axes=True, axes_labels=[x,y],color="red")
```

## Поверхность:
Задана следующая квадратичная форма:
    $$ 8x^2+2y^2+20z^2+16xy+16xz+32yz $$

```python
#уравнение к.ф. в символьном виде
var('x y z')
g(x,y,z) = 8*x^2 + 2*y^2 + 20*z^2 + 16*x*y + 16*x*z + 32*y*z
```

Теперь построим график заданной поверхности:

```python
implicit_plot3d(g, (-20,20), (-20,20), (-20,20), color="red")
```

Составим для данной поверхности матрицу $Q$ и вектор $v$.

```python
Qd = matrix([[8,8,8],[8,2,16],[8,16,20]])
vd = matrix([x,y,z])
Qd
```

Теперь нам нужно диагонализировать матрицу КФ: для этого нам нужно перейти из стандартного базиса к базису из собственных векторов. Следовательно, следует найти собственные числа и векторы матрицы $Q$:

```python
vecd3, vecd2, vecd1 = Qd.right_eigenvectors() #получаем тройки значений и записываем их в переменные
vecd1 = vector(vecd1[1][0]) #получаем собственный вектор1
vecd2 = vector(vecd2[1][0]) #получаем собственный вектор2
vecd3 = vector(vecd3[1][0]) #получаем собственный вектор1
print("первый вектор: " + str(vecd1) + "\n")
print("второй вектор: " + str(vecd2) + "\n")
print("третий вектор: " + str(vecd3) + "\n")
```

Получив собственные векторы матрицы можно задать матрицу перехода от стандартного базиса в $R^3$ к базису из собственных векторов. Запишите матрицу перехода, составив ее из собственных векторов и при необходимости избавившись от дробей в векторах (вектора следует нормировать):

```python
new_matrixd = matrix([vecd1, vecd2, vecd3])
print(new_matrixd)
```

Теперь проверьте равенство норм обоих собственных векторов (обратите внимение, здесь используется стандартная евклидова, т.е. вторая, норма):

```python
if (vecd1.normalized().norm(2) == vecd2.normalized().norm(2)) and (vecd1.normalized().norm(2) == vecd3.normalized().norm(2)) and (vecd2.normalized().norm(2) == vecd3.normalized().norm(2)):
    print("Равенство норм верно")
else:
    print("Равенство норм не верно")
```

Перепишем теперь матрицу $Q$ в новом базисе:

```python
new_Qd = new_matrixd*Qd*new_matrixd^(-1)
print(new_Qd)
```

Теперь подставим полученные переменные в наше исходное уравнение:

```python
equationd = (vd*new_Qd*vd.transpose())[0][0]
print("Новое полученное уравнение: " + str(equationd))
```

Построим график, который после поворота уже оцентрован:

```python
implicit_plot3d(equationd, (-30, 30), (-30,30), (-30, 30), color="red")
```
