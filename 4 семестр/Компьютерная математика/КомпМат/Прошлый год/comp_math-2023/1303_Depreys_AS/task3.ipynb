## Основные теоретические положения

### Симметрическая группа сттепени $n$
Пусть $M = \left\{1,2,\dots,n\right\}$ - конечное множество из $n$ элементов. Группа $S_n = S\left(M\right)$ всех взимно однозначных отображений мн-ва $M$ на себя с операцией умножения (композиции) называется симметрической группой степени $n$
1. Элементы группы называются подстановками.<br>
2. $\#S_n = n!$<br>
3. Действие подстановки $\sigma$ на множестве удобно описывать, указывая образы всех элементов множества $M$:
$$\sigma = \begin{pmatrix} 1 & 2 & \dots & n \\ i_1 & i_2 & \dots & i_n \end{pmatrix},\quad\text{где  } i_k = \sigma(i), \quad k = 1,2,\dots,n$$<br>
4. Поскольку подстановки биективны, то все $i_k$ - различны и исчерпывают все множество $M$.<br>
5. Подстановки $\sigma, \tau \in S_n$ умножаются в соответствии с общим правилом композиции отображений.<br>
6. Нейтральным элементом в группе является тождественная подстановка $e(i) = i,\; i = 1,2,\dots,n$<br>
7. Обратным элементомпо отношению к подстановке $\sigma $ служит обратное отображение $\sigma^{-1}$, для которого $i = \sigma^{-1}(i_k), \; k = 1,2,\dots,n$


### Цикл длины $k$
Подстановка является циклом длины $k$, если на некотором подмножестве $\left\{a_{i_1},a_{i_2},\dots,a_{i_k}\right\}$ множества $M$ она действует следующим образом:$$\begin{pmatrix} a_{i_1} & a_{i_2} & \dots & a_{i_{k - 1}} & a_{i_k} \\ a_{i_2} & a_{i_3} & \dots & a_{i_k} & a_{i_1} \end{pmatrix},$$ а остальные элементы множества $M$ не меняет.
1. Цикл принято обозначать $\left(a_{i_1},a_{i_2},\dots,a_{i_k}\right)$<br>
2. Два цикла являются независимыми, если действуют на непересекающихся подмножествах множества $M$.<br>
3. Пусть $\sigma$ - цикл длины $k$. Тогда он порождает в $S_n$ циклическую подгрупу $\left<\sigma\right> = \left\{e, \sigma,\sigma^{2},\dots,\sigma^{k - 1}\right\}$ порядка $k$.<br>
$\left(\sigma^{s}\right)^{-1} = \sigma^{k - s}, \quad s = 1,2,\dots,k-1$


### Теорема
Каждая подстановка $\sigma \ne e \text{ в } S_n$ является произведениеем независимых циклов. Это разложение в произведение определено однозначно с точностью до порядка следования циклов.
### Следствие
Порядок подстановки $\sigma = $ наименьшему общему кратному длин независимых циклов, входящих в разложение $\sigma$. Порядок подстановки является порядком циклической подгруппы $\left<\sigma\right>$

### Действие группы на множестве
Пусть $G$ - группа с нейтральным элементом $e$, а $X$ - множество. Будем говорить, что $G$ действует на $X$, если задана операция $G\times X \to X$, обладающая для любого $x \in X$ и $g,h \in G$ следующими свойствами:
1. $ex = x$<br>
2. $gh(x) = g\left(h(x)\right)$

### Орбита
Орбитой элемента $x \in X$ под действием $G$ называется множество $Gx = \left\{gx \; :g \in G\right\}$. Число $\#\left(Gx\right)$ называется длиной орбиты.

### Неподвижные точки
Неподвижными точками элемента $g \in G$ называются точки $x \in X$, для которых $g(x) = x$. Множество неподвижных точек элемента $g$ обозначается $X^g$


### Лемма Бернайда
Количество орбит действия группы $G$ на множестве $X$ равно:$$N = \frac{1}{\#G}\cdot \sum_{g \in G}\# X^g$$

### Раскраска множества
Пусть $I$ - произвольное множество, а $C$ - множество цветов. Раскраской множества $I$ называется функция из $I$ в $C$. Множество всех раскрасок обозначается как $C^{I}$. Количество раскрасок равно $\left(\# C\right)^{\# I}$

### Лемма 
Пусть $G$ - группа, действующая на множестве $I$. Тогда формула $gf(i) = f\left(g^{-1}i\right)$, где $g \in G, i\in I, f\in C^{I}$, задает действие $G$ на $C^{I}$

### Лемма 
Для элемента $g \in G$ и функции $f \in C^{I}$ следующие условия эквивалентны:
1. $gf = f$<br>
2. $f(i) = f(g^{n}i)$ для всех $i \in I$ и $n \in \mathbb{Z}$<br>
3. Если $i,j \in I$ лежат в одном цикле в циклической записи перестановки $\varphi(g)$, то $f(i) = f(j)$

### Следствие
Количество раскрасок из $C^{I}$, которые сохраняет данный элемент $g \in G$ равно $\left(\#C\right)^{c\left(\varphi(g)\right)}$. Здесь $c(\sigma)$ обозначает количество независимых циклов в записи перестановки $\sigma$

### Лемма
Предположим $\sigma$ является циклом длины $k$, а $\gcd(k,n) = d$. Тогда перестановка $\sigma^{n}$ является произведением $d$ циклов длины $\frac{k}{d}$

```python
def coloring_by_num(n):
    if n < 0 or n > 244140624:
        return
    colors = ['red', 'yellow', 'blue', 'green', 'brown']
    coloring = [0 for i in range(12)]
    for i in range(12):
        coloring[i] = n // pow(5,11 - i);
        n %= pow(5,11 - i)
        
    Edges = Graphics()
    Edges += line3d(((0, 0, 0),(1, 0, 0)), viewer='threejs', thickness = 5, color = colors[coloring[0]])
    Edges += line3d(((1, 0, 0),(1, 0, 1)), viewer='threejs', thickness = 5, color = colors[coloring[1]])
    Edges += line3d(((1, 0, 1),(0, 0, 1)), viewer='threejs', thickness = 5, color = colors[coloring[2]])
    Edges += line3d(((0, 0, 1),(0, 0, 0)), viewer='threejs', thickness = 5, color = colors[coloring[3]])
    Edges += line3d(((1, 0, 0),(1, 1, 0)), viewer='threejs', thickness = 5, color = colors[coloring[4]])
    Edges += line3d(((1, 0, 1),(1, 1, 1)), viewer='threejs', thickness = 5, color = colors[coloring[5]])
    Edges += line3d(((0, 0, 1),(0, 1, 1)), viewer='threejs', thickness = 5, color = colors[coloring[6]])
    Edges += line3d(((0, 0, 0),(0, 1, 0)), viewer='threejs', thickness = 5, color = colors[coloring[7]])
    Edges += line3d(((0, 1, 0),(1, 1, 0)), viewer='threejs', thickness = 5, color = colors[coloring[8]])
    Edges += line3d(((1, 1, 0),(1, 1, 1)), viewer='threejs', thickness = 5, color = colors[coloring[9]])
    Edges += line3d(((1, 1, 1),(0, 1, 1)), viewer='threejs', thickness = 5, color = colors[coloring[10]])
    Edges += line3d(((0, 1, 1),(0, 1, 0)), viewer='threejs', thickness = 5, color = colors[coloring[11]])
                
    Edges += text3d("1", (0.5, 0, 0), fontsize=40)
    Edges += text3d("2", (1, 0, 0.5), fontsize=40)
    Edges += text3d("3", (0.5, 0, 1), fontsize=40)
    Edges += text3d("4", (0, 0, 0.5), fontsize=40)
    Edges += text3d("5", (1, 0.5, 0), fontsize=40)
    Edges += text3d("6", (1, 0.5, 1), fontsize=40)
    Edges += text3d("7", (0, 0.5, 1), fontsize=40)
    Edges += text3d("8", (0, 0.5, 0), fontsize=40)
    Edges += text3d("9", (0.5, 1, 0), fontsize=40)
    Edges += text3d("10", (1, 1, 0.5), fontsize=40)
    Edges += text3d("11", (0.5, 1, 1), fontsize=40)
    Edges += text3d("12", (0, 1, 0.5), fontsize=40)
    Edges.show(frame = False)
    print(coloring)
```
