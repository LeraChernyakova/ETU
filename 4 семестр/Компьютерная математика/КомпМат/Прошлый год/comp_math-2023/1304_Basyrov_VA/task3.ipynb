## Основные теоретические положения

### Действие группы на множество
Пусть $G$ - группа с нейтральным элементом $e$, a $X$ - множество. Будем говорить, что $G$ действует на $X$, если задана операция $G \times X \rightarrow X$.   
$\forall$ $x \in X$ и $g, h \in G$, при том:
<ol>
<li>$ex = x$</li>
<li>$g(h(x)) = gh(x)$</li>
</ol>


### Орбита
Орбитой элемента $x \in X$ под действием $G$ называется множество  
$Gx = \{gx : g \in G\}$
Длина орбиты - количество элементов в ней, обозначается $\#G$

### Неподвижные точки
Неподвижными точками элемента $g \in G$ называются $x \in X : gx = x$<br>
Множество неподвижных точек элемента $g$ обозначается $X^g$

### Лемма Бернсайда
Количество орбит действия группы $G$ на множестве $X$ равно  
$\frac{1}{\#G}\sum_{g \in G} \#X^{g}$

### Определение раскраски
Пусть $I$ - произвольное множество, а $C$ - множество цветов. Раскраской множества $I$ называется функция из $I$ в $C$. Множество всех раскрасок обозначается через $C^I$, количество раскрасок равно $\#C^{\#I}$ 

### Количество раскрасок
Количество раскрасок из $C^I$, которые сохраняют данный элемент $g \in G$ равно $\#C^{c(\varphi(g)))}$  
$с(\varphi(g))$ - количество независимых циклов перестановки $\varphi(g)$

### 1)Аналитически решить задачу о раскрасках для заданного варианта.

Создадим все необходимые симметричные группы, которые далее будем использовать для решения поставленной задачи.

```python
Base_Sym_G=SymmetricGroup(62)
Default_Sym_G=Base_Sym_G(tuple(srange(1,63)))
Reflection_Sym_G=Base_Sym_G(srange(62,0,-1))
Neutral_Sym_G = Base_Sym_G([(i) for i in range(1, 63, 1)])
```

Выведем степени перестановки "по умолчанию"(1,2,...,62), а также степени этой перестановки умноженную на перестановку отражения ((1,62),(2,61),...,(30,31))

```python
for power in range(1,62):
    print(f"\u03C4^{power} =",Default_Sym_G**power)
```

```python
for power in range(1,62):
    print(f"\u03C3 * \u03C4^{power} =", (Default_Sym_G**power)*Reflection_Sym_G)
```

Используя полученные данные, по лемме Бернсайда найдем количество орбит для заданного количество цветов. Это и будет ответом на исходную задачу.
Зададим формулой действие элементов группы G на множестве X с помощью $ G^{*} $:
$$ g(\{x_1, x_2, x_3, x_4, ..., x_{62}\}) = \{x_{(g^{*})^{−1}(1)}, x_{(g^{*})^{−1}(2)}, x_{(g^{*})^{−1}(3)}, \dots, x_{(g^{*})^{−1}(62)}\}, \text{ где } g^{*} \in G^{*}, g \in G $$
В соответствии с леммой Бернсайда, количество орбит действия группы $ G $ на множестве $ X $ равно:
$$N = \frac{1}{\#G}\sum_{g \in G}{\#X^g}$$
А орбиты действия - это разбиение множества X на классы эквивалентности относительно действий G, это искомое число способов составить бусы с точностью до поворотов и осевых симметрий.<br>
Найдем $ \#X^g $ (количество раскрасок, которые остаются на месте под действием преобразования g) для каждого $ g \in G $:
$ \tau, \tau^2, ..., \sigma \times \tau^{61} $ состоят из произведений независимых циклов. Получаем, что если все бусины в одном цикле одинакого цвета, тогда под действием преобразования сама расцветка бус не изментся, изменятся лишь позиции, а это и есть $ X^g $. <br>
Тогда $ \#X^g = 4^{сycles(g)} $, где $ сycles(g) $ - количество независимых циклов в перестановке $ g $, а 4 - количество цветов.<br>
Посчитаем $\#X^g $ для всех $ g \in G$.<br>
Для этого была написана программа, представленная ниже. Она также считает $\sum_{g \in G}{\#X^g}$ и $N$.

```python
different_colors=4
sum_coloring=0
for power in range(1, 62):
    X_g = len((Default_Sym_G**power).cycle_tuples(singletons=True))
    sum_coloring += different_colors ** X_g
    print(f"#X^(\u03C4^{power}) = {different_colors}^{X_g} =", different_colors ** X_g)
X_g = len(Neutral_Sym_G.cycle_tuples(singletons=True))
sum_coloring += different_colors ** X_g
print(f"#X^e = {different_colors}^{X_g} =", different_colors ** X_g)
X_g = len(Reflection_Sym_G.cycle_tuples(singletons=True))
sum_coloring += different_colors ** X_g
print(f"#X^(\u03C3) = {different_colors}^{X_g} =", different_colors ** X_g)
for power in range(1,62):
    X_g = len(((Default_Sym_G**power)*Reflection_Sym_G).cycle_tuples(singletons=True))
    sum_coloring += different_colors ** X_g
    print(f"#X^(\u03C3 * \u03C4^{power}) = {different_colors}^{X_g} =", different_colors ** X_g)
print("\nИтог:\n#X^g=",sum_coloring)
print("N=",sum_coloring/124)
```

### 2)Реализовать и продемонстировать функцию отображение раскраски по заданному номеру

```python
def rotate(position, angle):
    angle = angle * pi()/180
    return (matrix([[cos(angle), -sin(angle)],[sin(angle), cos(angle)]]) * position.transpose()).transpose()


def show_beads(coloring, distance = 300, size = 100, beads = 62):
    colors = ["red", "green", "blue", "yellow"]
    beads_color = []
    for i in range(62):
        beads_color.append(colors[coloring % different_colors])
        coloring //= different_colors
    position = matrix([distance, 0])
    base_angle = 360 / beads
    bus = point(position.list(), color = beads_color[0], size = size)
    for i in range(1, beads):
        center_position = (rotate(position, base_angle * i)).list()
        bus += point(center_position, color = beads_color[i], size = size)
    show(bus, axes=False)
```
