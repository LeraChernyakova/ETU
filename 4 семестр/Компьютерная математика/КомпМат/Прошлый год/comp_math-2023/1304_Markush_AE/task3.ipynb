## Основные теоретические положения

### Определение группы

Группой называется множество G с заданной бинарной операцией $*$, если выполнены следующие аксиомы:
<ol>
    <li>Множество замкнуто относительно операции $*$.</li>
    <li>Ассоциативность, т.е. $ (a * b) * c = a * (b * c)   \forall a, b, c \in G $.</li>
    <li>Наличие нейтрального элемента $ e \in G $ такого, что $ a * e = e * a = a, \forall a \in G $.</li>
    <li>Наличие обратных элементов такого, что $ \forall a \in G $ $ \exists a^{-1} \in G $: $ a^{-1} * a = a * a^{-1} = e $.</li>
</ol>
Пусть $ G $ - группа с нейтральным элементом e, $ X $ - какое-то множество. Говорим, что G действует на X, если задана операция $ G \times X \rightarrow G $ такая, что $ \forall x \in X $ и $ g, h \in G $ справедливо:
<ol>
    <li>$ e(x) = x $</li>
    <li>$ g \times h(x) = g(h(x)) $</li>
</ol>
Орбитой элемента $x \in X $ под действием $ G $ называется множество $ G_x = \{ gx: g \in G \} $. Число $ \#(G_x) $ называется длиной орбиты.<br>
Неподвижными точками элемента $ g \in G $ называется $ x \in X $, для которых $ gx = x $. Множество неподвижных точек элемента $ g $ обозначают $ X^g $.

### Лемма Бернсайда

Количество орбит действия группы $ G $ на множестве $ X $ равно:
$$
    N = \frac{1}{\#G}\sum_{g \in G}{\#X^g}
$$

### Определение раскраски
 
Пусть $I$ - произвольное множество, $С$ - множество цветов. Раскраской множества $I$ называется функция $I \rightarrow C$
 
Множество всех раскрасок: $C^I$.

Их количество: $\#C^I = \#C^{\#I}$

Лемма. Пусть группа  G  действует на множество  I , тогда формула $$gf(i)=f(g^{−1}(i)),g \in G,i \in I,f \in C^I$$
задаёт действие $G$ на множество  $C^I$.

Для $g \in G$ и $f \in C^I$ справедливо:
<ol>
<li>$gf = f$</li>

<li>$f(i) = f(g^{n}i)$, $\forall i \in I $ и $n \in \mathbb{Z}$</li>

<li>Если $i,j \in I$ лежат в одном цикле в циклической записи перестановки $\varphi(g)$, то $f(i) = f(j)$</li>
</ol>

Количество раскрасок из множетсва $C^I$, которые сохраняют элемент $g \in G$ равно $\#C^{c(\phi(g))}$, где $c(\phi(g))$ - число независимых циклов перестановки $\phi(g)$

### Задача
Сколькими способами можно составить бусы из 62 бусинок 3 различных цветов (с точностью до поворотов и осевых симметрий)?

### Аналитическое решение задачи:

Пусть X - множество всех раскрасок, которые записываются в виде $ \{ x_1, x_2, x_3, ..., x_{62} \} $, где $ x_k = \{ a, b, c \} $ - один из 3-ёх цветов.<br>
Тогда всего (с учетом поворотов и осевых симметрий) в множестве $ X $ - $ 3^{62} $ элементов.<br>
Пусть G - группа поворотов и осевых симметрий множества X, она изоморфна подгруппе $ G^* \in S_{62} $,  $ G^{*} = \{ 1, 2, 3, 4, ..., 62 \} $<br>
Рассмотрим элементарные преобразования положений бусин:<br>
Пусть $ \tau $ - поворот на 1 бусину по часовой стрелке, а $ \tau $ - осевая симметрия относительно оси, отделяющей $ \{ 1, 2, 3, ..., 30 \} $ от $ \{ 31, 32, 33, ..., 62 \} $.<br>
Опишем все возможные повороты и осевые симметрии для бус.<br>
Поворотот на 1 бусину по часовой:
$$ \tau = (1, 2, 3, 4, ... 61, 62) $$
Всего возможно 61 поворота, на 62 поворот бусы встанут в изначальное положение.

```python
G = SymmetricGroup(62)
s = G(tuple(srange(1, 63)))
for i in range(1, 62):
    print(f"\u03C4^{i} =", s**i)
```

Отражение бус по оси симметрии:$$\sigma = (1, 62)(2, 61)(3, 60)\dots(28, 33)(29, 32)(30, 31)$$

Ниже представлне код, считающий всевозможные комбинации осевой симметрии и поворотов:

```python
G = SymmetricGroup(62)
s = G(tuple(srange(1, 63)))
t = G(srange(62, 0, -1))
for i in range(1, 62):
    print(f"\u03C3 * \u03C4^{i} =", (s**i)*t)
```

Зададим формулой действие элементов группы G на множестве X с помощью $ G^{*} $:
$$ g(\{x_1, x_2, x_3, x_4, ..., x_{62}\}) = \{x_{(g^{*})^{−1}(1)}, x_{(g^{*})^{−1}(2)}, x_{(g^{*})^{−1}(3)}, \dots, x_{(g^{*})^{−1}(62)}\}, \text{ где } g^{*} \in G^{*}, g \in G $$
Например, цвет первой бусины $ x_1 $ после поворота по часовой на одну бусину ($ \tau $) стал $ x_{62} $, что соответствует $ \tau^{-1}(1) = \tau(61) = 62 $.<br>
В соответствии с леммой Бернсайда, количество орбит действия группы $ G $ на множестве $ X $ равно:
$$N = \frac{1}{\#G}\sum_{g \in G}{\#X^g}$$
А орбиты действия - это разбиение множества X на классы эквивалентности относительно действий G, это искомое число способов составить бусы с точностью до поворотов и осевых симметрий.<br>
Найдем $ \#X^g $ (количество раскрасок, которые остаются на месте под действием преобразования g) для каждого $ g \in G $:
$ \tau, \tau^2, ..., \sigma \times \tau^{61} $ состоят из произведений независимых циклов. Получаем, что если все бусины в одном цикле одинакого цвета, тогда под действием преобразования сама расцветка бус не изментся, изменятся лишь позиции, а это и есть $ X^g $. <br>
Тогда $ \#X^g = 3^{сycles(g)} $, где $ сycles(g) $ - количество независимых циклов в перестановке $ g $, а 3 - количество цветов.<br>
Посчитаем $\#X^g $ для всех $ g \in G$.<br>
Для этого была написана программа, представленная ниже. Она также считает $\sum_{g \in G}{\#X^g}$ и $N$.

```python
G = SymmetricGroup(62)
e = G([(i) for i in range(1, 63, 1)])
s = G(tuple(srange(1, 63, 1)))
t = G(srange(62, 0, -1))
summ = 0

X_g = len(e.cycle_tuples(singletons=True))
summ += 3 ** X_g
print(f"#X^e = 3^{X_g} =", 3 ** X_g)

X_g = len(s.cycle_tuples(singletons=True))
summ += 3 ** X_g
print(f"#X^\u03C4 = 3^{X_g} =", 3 ** X_g)
for i in range(2, 62):
    X_g = len((s**i).cycle_tuples(singletons=True))
    summ += 3 ** X_g
    print(f"#X^(\u03C4^{i}) = 3^{X_g} =", 3 ** X_g)
    
X_g = len(t.cycle_tuples(singletons=True))
summ += 3 ** X_g
print(f"#X^(\u03C3) = 3^{X_g} =", 3 ** X_g)
for i in range(1,62):
    X_g = len(((s**i)*t).cycle_tuples(singletons=True))
    summ += 3 ** X_g
    print(f"#X^(\u03C3 * \u03C4^{i}) = 3^{X_g} =", 3 ** X_g)
    

print(f"#X^g sum =", summ)
print(f"N=", summ/124)
```

Получаем искомое число возможных раскрасок бус из 62 бусинок 3 цветов (с точностью до поворота и осевой симметрии):

$$ N = 3076777616750185812928914456 $$

### Графическое отображение раскраски по заданному номеру:

```python
def rotate(x_y, angle):
    angle = angle * pi()/180
    return (matrix([[cos(angle), -sin(angle)],[sin(angle), cos(angle)]]) * x_y.transpose()).transpose()


def paint_beads(coloring, distance = 200, size = 100, circles = 62):
    num = 3
    colors = ["palegreen", "wheat", "limegreen", "plum", "gainsboro", "deeppink"]
    circles_color = []
    for _ in range(62):
        circles_color.append(colors[coloring % num])
        coloring //= num
    x_y = matrix([distance, 0])
    base_angle = 360 / circles
    bus = point(x_y.list(), color = circles_color[0], size = size)
    for i in range(1, circles):
        curr_center_x_y = (rotate(x_y, base_angle * i)).list()
        bus += point(curr_center_x_y, color = circles_color[i], size = size)
    show(bus, axes=False)
```

```python
paint_beads(0)
```

```python
paint_beads(11)
```
