## Основные теоретические положения

Аддитивной цепочкой для натурального числа $n$ называют последовательностью натуральных чисел:  
<br>
\begin{align}
    1 = a_0, a_1, a_2, … , a_r = n  
\end{align}
<br>
где каждый элемент последовательности равен сумме каких-либо двух предыдущих:   
<br>
\begin{align}
    a_i = a_j + a_k \qquad k \leqslant j < i \qquad i = 1,2,3,…,r
\end{align}
<br>

### Аддитивное дерево
* $l(n) = r$ - наименьшая длина $r$, для которой существует аддитивная цепочка для $n$
* Неравенство для метода множителей:
<br>
\begin{align}
    l(mn) \leqslant l(m) + l(n)
\end{align}
<br>
* m - равный метод $ m=2^k, n = \sum_{j = 0}^t d_j m^{t-j}$  
<br>
\begin{align}
		l(n) \leqslant m - 2 + (k + 1)t
\end{align}
<br>
* Бинарный метод $“SX”$:
<br>
\begin{align}
    l(n) \leqslant \lambda(n) + \nu(n) - 1
\end{align}
<br>
* Полагаем, что все аддитивные цепочки возрастают:
<br>
\begin{align}
    1 = a_0 < a_1 < … < a_r = n
\end{align}
<br>
* Если два числа из {$a_i$} одинаковые, то одно из них может быть опущено
* Пара $(j,k)$, $0 \leqslant k \leqslant j < i$ называется шагом $i$
* Если существует более чем одна пара $(j,k)$ полагаем, что $j$ - наибольший из возможных

#### Виды:
* Удвоение: $j = k = i - 1$
* Звездный шаг: $j = i - 1$
* Малый шаг: $\lambda(a_i) = \lambda(a_{i-1})$

#### Свойства:
* Шаг 1 всегда удвоение
* Удвоение - звездный шаг, но никогда не малый шаг 
* За удвоением всегда следует звездный шаг
* Если $i$ - ый шаг не малый, то $(i + 1)$ - ый шаг либо малый, либо звездный, либо и тот и другой
* Если $(i + 1)$ - ый шаг ни звездный, ни малый, то $i$ - ый шаг должен быть малым 


### Теорема:
Если аддитивная цепочка включает $d$ удвоений и $f = r - d$ неудвоений, то $n \leqslant 2^{d-1}F_{f + 3}$, где $F_j$ - число Фибоначи

#### Следствие:
Если аддитивная цепочка включает $f$ неудвоений и $s$ малых шагов, то
<br>
\begin{align}
    s \leqslant f \leqslant \frac{s}{1 - log_2 \phi}
\end{align}
<br>    
где $\phi = \frac{\sqrt{5} + 1}{2}$ - золотое сечение 

<br>
Алгоритм Брауэра позволяет вычислить $n$ - ую степень за  
<br>
\begin{align}
    l_B(n) = \lambda(n)+ \frac{\lambda(n)}{\lambda (\lambda(n))} + O \Bigr(\frac{\lambda(n)\lambda(\lambda (\lambda(n))))}{(\lambda (\lambda(n)))^2}\Bigl) 
\end{align}
<br> 
умножений

* Самая короткая аддитивная цепочка для числа $n$ имеет длину не более $\lambda(n)$ 
* Доказано, что почти для всех $n$ минимальная аддитивная цепочка имеет длину $l_B(n)$

<br>

Для натурального числа $n$ при заданном натуральном числе $k$ можно построить цепочку Брауэра с помощью рекуррентный формы:  

\begin{align}
B_k (n) = \left\{
\begin{array}{cl}
1, 2, 3, ..., 2^k - 1 & n < 2^k \\
B_k(q), 2q, 4q, 8q, ..., 2^kq, n & n \geq 2^k, q = \lfloor\frac{n}{2^k}\rfloor.
\end{array}
\right.
\end{align}


* Данная цепочка будет иметь длину:
<br>
\begin{align}
    l_B(n) = j(k + 1) + 2^k - 2
\end{align}
<br>
при условии, что $j_k\leqslant \lambda(n) < (j + 1)k$
* Длина будет минимизирована для больших $n$, если положить $k = \lambda(\lambda(n)) - 2\lambda(\lambda(\lambda(n)))$

### Алгоритм Брауэра

1. Задаётся некий фиксированный параметр $к$ для рассматриваемого числа $n$. Выполняется вычисление “вспомогательных чисел”:  
\begin{array}{cl}
d = 2^k, q_1 = [\frac{n}{d}],& r_1 = n  mod  d \Rightarrow n = q_1d + r_1 & (0 \leqslant r_1 < d  ) \\
q_2 = [\frac{q_1}{d}],& r_2 = q_1  mod  d \Rightarrow q_1 = q_2d + r_2  &(0 \leqslant r_2 < d  )
\end{array}
2. Данная процедура продолжается до тех пор, пока не появится такое $q_s < d$. Следовательно $q_{s-1} = q_sd+r_s$
3. Таким образом $n$ имеет вид: 
<br>
\begin{align}
    n = 2^kq_1 + r_1 = 2^k(2^kq_2 + r_2) + r_1 = ... = 2^k(2^k(...(2^kq_s + r_s) ... )+ r_2)+ r_1
\end{align}
<br>

### Алгоритм дробления вектора индексов 
1. Внешний цикл по длинам цепочек: $\underline{l}(n) \leqslant m \leqslant\overline{l}(n)$. Выбирается число $q \in \mathbb{N}$. Пусть $q = \frac{m}{2}$
2. Внутренний цикл перебора всех ${r_1, r_2, …, r_q}$ ($q!$   шагов). На каждом шаге при фиксированной части вычисляем $a_{min}$ и $a_{max}$:
    1. Если $a_m = n$ - задача решена
    2. Если $n \notin [a_{min},a_{max}]$, то переходим к следующему набору ${r_1, r_2, ..., r_q}$
    3. Если $n \in [a_{min},a_{max}]$, то организуем внутренний цикл перебора меняющейся части ${\rho_{q + 1}, \rho_{q + 2},..., \rho_m}$. Таких наборов $\frac{m!}{q!}$ штук:
        1. Если обнаружен $a_m = n$ - задача решена
        2. Если в цикле таких векторов не оказалось, то переходим к следующей (по введенной упорядочености) фиксированой части ${r_1, r_2, ..., r_q}$.
3. Если все наборы фиксированной длины исчерпаны, то увеличиваем их длину во внешнем цикле.

#### Предположения
1. Для чисел $n < 12509$ справедливо $l(n) = l^*(n)$
2. $l(2n) = l(n)$: Минимальное число $n = 191$
3. $l(2n) < l(n)$: минимальное число $n = 375494703$
4. Гипотеза Шольца - Брауэра: $l(2^n-1) \leqslant l(n) + n - 1$
    1. Гипотеза доказана для звездных цепочек: $l^*(2^n-1) \leqslant l^*(n) + n - 1$
    2. Гипотеза справедлива для всех $n < 5784689$
    3. Равенство выполняется для $1 \leqslant n \leqslant 64$

### Алгоритм Брауэра
##### Задача: 
Реализовать алгоритм Брауэра для вычисления приближённых аддитивных цепочек для различных чисел при варьировании параметра 𝑘, сопоставить длины полученных аддитивных цепочек с минимальной аддитивной цепочкой для заданного числа. Сделать выводы.


```python

#алгоритм Брауэра
def brauer_algorithm(n, k):
    
    # переменная для хранения аддитивной цепочки
    additive_chain = [] 
    d = 2**k 
    
    # начальная инициализация цепочки
    for i in range(1,d):
        additive_chain.append(i)
    
    list_q = []
    # вспомогательные числа
    list_r = [] 
    q = n // d
    r = n % d
    list_r.append(r)
        
    # заполнение массива вспомогательных чисел
    while not q < d:
        r = q % d
        q = q // d
        list_r.append(r)
        
    list_r.append(q)
        
    #переменная для вычисления элемента через вспомогательное число
    tmp_need = list_r[-1]
    
    # обработка случая когда d > n (n будет в цепочке при начальной инициализации)
    if n in additive_chain:
        return additive_chain[:n]
    
    # вычисление элементов цепочки через вспомогательные числа
    while additive_chain[-1]!=n:
        
        while not q == tmp_need * d:
            q += q
            if(q not in additive_chain):
                additive_chain.append(q) 
            
        list_r.pop()
        
        tmp_need = q + list_r[-1]
        q = tmp_need
        
        # если элемент уже есть в цепочке, то он не добавляется (так как в цепочке элементы не повторяются)
        if(q not in additive_chain):
            additive_chain.append(tmp_need)
          
    
    return additive_chain


# result=brauer_algorithm(7182, 3)

# print(result)
# print(len(result))
```

#### Задача
Реализовать алгоритм дробления вектора индексов для нахождения минимальной звёздной цепочки для заданного числа. Протестировать алгоритм минимум для 5 значений 𝑛 > 1000. Указать, сколько времени потребовалось на поиск цепочки и какая цепочка получилась. Сравнить с предыдущими методами, сделать выводы.

```python
#функция для добавления единицы в вектор индексов
def change_chain(list_ind, q):
    for i in range(len(list_ind)):

        if list_ind[len(list_ind) - i - 1] == 1: 
            list_ind[len(list_ind) - i - 1] = len(list_ind) - i + q
        else:
            list_ind[len(list_ind) - i - 1] = list_ind[len(list_ind) - i - 1] - 1
            break
    return list_ind
        
#алгоритм дробления индексов        
def crushing_indexes(n):
    #определение границ m
    min_m = floor(log(n,2)) + floor(log(str(bin(n)).count('1'),2))
    max_m = floor(log(n,2)) + str(bin(n)).count('1') - 1
    
    #внешний цикл перебора m
    for m in range(min_m, max_m + 1):
        
        q = m // 2 
        #инициализация изменяемой и фиксированой части
        list_r = [i for i in range(1,q+1)]
        list_ro = [i for i in range (q+1, m+1)]
        
        #цикл перебора по list_r
        for step in range(factorial(q)+1):
        
            tmp_chain = [1]
            for ind in list_r+list_ro:
                tmp_chain.append(tmp_chain[-1] + tmp_chain[ind-1])
            
       
            a_min = tmp_chain[q] + m - q
            a_max = tmp_chain[q] * pow(2,m-q)
        
            
            if tmp_chain[-1] == n:
                return tmp_chain
            
            
            elif n < a_min or n > a_max :
                list_r = change_chain(list_r,0)
                
                
            else:
                #внутренний цикл перебора по list_ro 
                for inner_step in range(factorial(m)//factorial(q)+1):
                    
                    tmp_chain = [1]
                    for ind in list_r+list_ro:
                        tmp_chain.append(tmp_chain[-1] + tmp_chain[ind-1])
                    
                    if tmp_chain[-1] == n:
                        return tmp_chain
                    
                    list_ro = change_chain(list_ro,q)
                    
                list_r = change_chain(list_r,0)
                list_ro = [i for i in range (q+1, m+1)]


# print("START")
# print(crushing_indexes(16))
# print("END")
```
