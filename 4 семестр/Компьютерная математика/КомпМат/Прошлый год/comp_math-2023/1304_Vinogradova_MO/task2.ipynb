–†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º –ë—Ä–∞—É—ç—Ä–∞ –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ø—Ä–∏–±–ª–∏–∂—ë–Ω–Ω—ã—Ö –∞–¥–¥–∏—Ç–∏–≤–Ω—ã—Ö —Ü–µ–ø–æ—á–µ–∫ –¥–ª—è —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —á–∏—Å–µ–ª –ø—Ä–∏ –≤–∞—Ä—å–∏—Ä–æ–≤–∞–Ω–∏–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ ùëò, —Å–æ–ø–æ—Å—Ç–∞–≤–∏—Ç—å –¥–ª–∏–Ω—ã –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –∞–¥–¥–∏—Ç–∏–≤–Ω—ã—Ö —Ü–µ–ø–æ—á–µ–∫ —Å –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–π –∞–¥–¥–∏—Ç–∏–≤–Ω–æ–π —Ü–µ–ø–æ—á–∫–æ–π –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.


```python

#–∞–ª–≥–æ—Ä–∏—Ç–º –ë—Ä–∞—É—ç—Ä–∞
def brauer_algorithm(n, k):
    
    # –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –∞–¥–¥–∏—Ç–∏–≤–Ω–æ–π —Ü–µ–ø–æ—á–∫–∏
    additive_chain = [] 
    d = 2**k 
    
    # –Ω–∞—á–∞–ª—å–Ω–∞—è –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è —Ü–µ–ø–æ—á–∫–∏
    for i in range(1,d):
        additive_chain.append(i)
    
    list_q = []
    # –≤—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω—ã–µ —á–∏—Å–ª–∞
    list_r = [] 
    q = n // d
    r = n % d
    list_r.append(r)
        
    # –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –º–∞—Å—Å–∏–≤–∞ –≤—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª
    while not q < d:
        r = q % d
        q = q // d
        list_r.append(r)
        
    list_r.append(q)
        
    #–ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —ç–ª–µ–º–µ–Ω—Ç–∞ —á–µ—Ä–µ–∑ –≤—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ
    tmp_need = list_r[-1]
    
    # –æ–±—Ä–∞–±–æ—Ç–∫–∞ —Å–ª—É—á–∞—è –∫–æ–≥–¥–∞ d > n (n –±—É–¥–µ—Ç –≤ —Ü–µ–ø–æ—á–∫–µ –ø—Ä–∏ –Ω–∞—á–∞–ª—å–Ω–æ–π –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏–∏)
    if n in additive_chain:
        return additive_chain[:n]
    
    # –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Ü–µ–ø–æ—á–∫–∏ —á–µ—Ä–µ–∑ –≤—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω—ã–µ —á–∏—Å–ª–∞
    while additive_chain[-1]!=n:
        
        while not q == tmp_need * d:
            q += q
            if(q not in additive_chain):
                additive_chain.append(q) 
            
        list_r.pop()
        
        tmp_need = q + list_r[-1]
        q = tmp_need
        
        # –µ—Å–ª–∏ —ç–ª–µ–º–µ–Ω—Ç —É–∂–µ –µ—Å—Ç—å –≤ —Ü–µ–ø–æ—á–∫–µ, —Ç–æ –æ–Ω –Ω–µ –¥–æ–±–∞–≤–ª—è–µ—Ç—Å—è (—Ç–∞–∫ –∫–∞–∫ –≤ —Ü–µ–ø–æ—á–∫–µ —ç–ª–µ–º–µ–Ω—Ç—ã –Ω–µ –ø–æ–≤—Ç–æ—Ä—è—é—Ç—Å—è)
        if(q not in additive_chain):
            additive_chain.append(tmp_need)
          
    
    return additive_chain


# result=brauer_algorithm(7182, 3)

# print(result)
# print(len(result))
```

```python
#—Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –µ–¥–∏–Ω–∏—Ü—ã –≤ –≤–µ–∫—Ç–æ—Ä –∏–Ω–¥–µ–∫—Å–æ–≤
def change_chain(list_ind, q):
    for i in range(len(list_ind)):

        if list_ind[len(list_ind) - i - 1] == 1: 
            list_ind[len(list_ind) - i - 1] = len(list_ind) - i + q
        else:
            list_ind[len(list_ind) - i - 1] = list_ind[len(list_ind) - i - 1] - 1
            break
    return list_ind
        
#–∞–ª–≥–æ—Ä–∏—Ç–º –¥—Ä–æ–±–ª–µ–Ω–∏—è –∏–Ω–¥–µ–∫—Å–æ–≤        
def crushing_indexes(n):
    #–æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≥—Ä–∞–Ω–∏—Ü m
    min_m = floor(log(n,2)) + floor(log(str(bin(n)).count('1'),2))
    max_m = floor(log(n,2)) + str(bin(n)).count('1') - 1
    
    #–≤–Ω–µ—à–Ω–∏–π —Ü–∏–∫–ª –ø–µ—Ä–µ–±–æ—Ä–∞ m
    for m in range(min_m, max_m + 1):
        
        q = m // 2 
        #–∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –∏–∑–º–µ–Ω—è–µ–º–æ–π –∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–æ–π —á–∞—Å—Ç–∏
        list_r = [i for i in range(1,q+1)]
        list_ro = [i for i in range (q+1, m+1)]
        
        #—Ü–∏–∫–ª –ø–µ—Ä–µ–±–æ—Ä–∞ –ø–æ list_r
        for step in range(factorial(q)+1):
        
            tmp_chain = [1]
            for ind in list_r+list_ro:
                tmp_chain.append(tmp_chain[-1] + tmp_chain[ind-1])
            
       
            a_min = tmp_chain[q] + m - q
            a_max = tmp_chain[q] * pow(2,m-q)
        
            
            if tmp_chain[-1] == n:
                return tmp_chain
            
            
            elif n < a_min or n > a_max :
                list_r = change_chain(list_r,0)
                
                
            else:
                #–≤–Ω—É—Ç—Ä–µ–Ω–Ω–∏–π —Ü–∏–∫–ª –ø–µ—Ä–µ–±–æ—Ä–∞ –ø–æ list_ro 
                for inner_step in range(factorial(m)//factorial(q)+1):
                    
                    tmp_chain = [1]
                    for ind in list_r+list_ro:
                        tmp_chain.append(tmp_chain[-1] + tmp_chain[ind-1])
                    
                    if tmp_chain[-1] == n:
                        return tmp_chain
                    
                    list_ro = change_chain(list_ro,q)
                    
                list_r = change_chain(list_r,0)
                list_ro = [i for i in range (q+1, m+1)]


# print("START")
# print(crushing_indexes(16))
# print("END")
```

```python

def crushing_indexes_test_hypothesis(n,m):   
    if n == 1:
        return [1]
    
    q = m // 2
    
    #–∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –∏–∑–º–µ–Ω—è–µ–º–æ–π –∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–æ–π —á–∞—Å—Ç–∏
    list_r = [i for i in range(1,q+1)]
    list_ro = [i for i in range (q+1, m+1)]
    
    #—Ü–∏–∫–ª –ø–µ—Ä–µ–±–æ—Ä–∞ –ø–æ list_r
    for step in range(factorial(q)+1):
        
        tmp_chain = [1]
        for ind in list_r + list_ro:
            tmp_chain.append(tmp_chain[-1] + tmp_chain[ind - 1])
            
        #print(tmp_chain)
        a_min = tmp_chain[q] + m - q
        a_max = tmp_chain[q] * pow(2,m-q)
        
            
        if tmp_chain[-1] == n:
            return tmp_chain
            
            
        elif n < a_min or n > a_max :
            list_r = change_chain(list_r,0)
                
                
        else:
            #–≤–Ω—É—Ç—Ä–µ–Ω–Ω–∏–π —Ü–∏–∫–ª –ø–µ—Ä–µ–±–æ—Ä–∞ –ø–æ list_ro
            for inner_step in range(factorial(m)//factorial(q)+1):
                    
                tmp_chain = [1]
                for ind in list_r+list_ro:
                    tmp_chain.append(tmp_chain[-1] + tmp_chain[ind-1])
                    
                if tmp_chain[-1] == n:
                    return tmp_chain
                    
                list_ro = change_chain(list_ro,q)
                    
            list_r = change_chain(list_r,0)
            list_ro = [i for i in range (q+1, m+1)]
    
        

def check_Scholz_Brauer_hypothesis():
    for n in range(1,13):
        second_statement = crushing_indexes(n)
        first_statement = crushing_indexes_test_hypothesis(2**n-1,len(second_statement) + n - 2)
        
        if len(first_statement) <=  len(second_statement) + n - 1:
            print(n,": l(2^n - 1) : ",first_statement, len(first_statement))
            print(n,": l(n) + n - 1 : ",second_statement, len(second_statement) + n - 1)
            print()
        else:
            print(n,": false", first_statement, second_statement)
            break
            
def print_table_Scholz_Brauer_hypothesis():
    print("|n|$l(2^n - 1)$|chain length|$l(n) + n - 1$|chain length|statement length|")
    print("|---|---|---|---|---|---|") 
    for n in range(1,13):
        second_statement = crushing_indexes(n)
        first_statement = crushing_indexes_test_hypothesis(2**n-1,len(second_statement) + n - 2)
        
        if len(first_statement) <=  len(second_statement) + n - 1:
            print("|",n,"|",first_statement,"|",len(first_statement),"|",second_statement,"|",len(second_statement),"|",len(second_statement) + n - 1,"|")
        else:
            
            break
check_Scholz_Brauer_hypothesis()
#print_table_Scholz_Brauer_hypothesis()
```
