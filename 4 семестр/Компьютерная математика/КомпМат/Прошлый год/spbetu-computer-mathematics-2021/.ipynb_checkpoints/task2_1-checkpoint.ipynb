## Основные теоретические положения
### Определение действия группы на множество.
Пусть $G$ - группа с нейтральным элементом $e$, a $X$ - множество. Будем говорить, что $G$ действует на $X$, если задана операция $G \times X \rightarrow X$.   
Для любого $x \in X$ и $g, h \in G$ операция обладает следующими свойствами:  
1) $ex = x$  
2) $g(hx) = (gh)x$


### Определение орбиты.
Орбитой элемента $x \in X$ под действием G называется множество  
$Gx = {gx | g \in G}$.  
Длина орбиты - количество элементов в ней. Любые две орбиты либо не пересекаются, либо совпадают.

### Определение неподвижных точек.
Неподвижными точками элемента $g \in G$ называются те $x \in X$, для которых $gx = x$.
Множество неподвижных точек элемента $g$ обозначается как $X^g$.

### Лемма Бернсайда.
Применяется в том случае, когда порядок множества $X$ намного больше порядка группы $G$.

Количество орбит действия группы G на множестве X равно  
$\frac{1}{|G|}\sum_{g \in G} |X^{g}|$

### Множество раскрасок.
Пусть I - произвольное множество, а C - множество цветов. Раскраской множества I называется функция из I в C. Множество всех раскрасок обозначается через $C^I$.  

### Количество раскрасок.
Количество раскрасок из $C^I$, которое сохраняют данный элемент $g \in G$ равно $|C|^{c\varphi(g))}$.  
$с(\varphi(g))$ - количество независимых циклов перестановки $\varphi(g)$ (считая циклы длины 1).

Группа вращений тетраэдра $G$ состоит из 12 вращений:

#### Вращения на угол $\pi$ вокруг осей, проходящих через середины противоположных ребер:

Вращение вокруг оси, соединяющей середины рёбер 1,2 и 3,4.  
$\begin{equation*}
{\sigma}_1 = \left(

Вращение вокруг оси, соединяющей середины рёбер 1,3 и 2,4.

Вращение вокруг оси, соединяющей середины рёбер 1,4 и 2,3.

\begin{array}{cccc}

\end{equation*}$ = (1 4)(2 3)

Группа $G$ действует на множестве раскрасок $X$. Найдем количество раскрасок. Количество цветов, которые надо выбрать для получения раскраски, сохраняемой элементом группы, равно количеству независимых циклов в перестановке, соответствующей этому элементу (при этом учитываются циклы длины 1). 

Нейтральный элемент оставляет на месте все раскраски:  
$|X^e| = 3^4$  
Вращение $\tau_1$ = (1)(2 4 3) сохраняет раскраску, если вершины 2, 3 и 4 окрашены в один и тот же цвет, цвет же первой вершины при этом может быть любым. Значит, первая вершина может быть окрашена 3 способами, и в то же время оставшиеся вершины красятся в один и тот же цвет.  
$|X^{\tau_1}| = 3\cdot3 = 3^2$  
Аналогично   
$|X^{\tau_1^2}| = |X^{\tau_2}| = |X^{\tau_2^2}| = |X^{\tau_3}| = |X^{\tau_3^2}| = |X^{\tau_4}| = |X^{\tau_4^2}| = 3^2$ 


Подстановка $\sigma_1$ = (1 3)(2 4) сохранит раскраску вершин при условии, что вершины 1 и 3 выкрашены в один цвет (3 возможности) и в то же время вершины 2 и 4 выкрашены в один цает (тоже 3 возможности). Тогда:  
$|X^{\sigma_1}| = |X^{\sigma_2}| = |X^{\sigma_3}| = 3^2$


Тогда количество раскрасок $N$ по лемме Бернсайда:  
$N = \frac{1}{|G|}\sum_{g \in G} |X^{g}| = \frac{1}{12}\cdot(3^4 + 11\cdot3^2) = 15$  
### Ответ: 15.

Функция рандомного распределения цветов вершинам тетраэдра:

```python
import random

def random1():
    vertexes = []
    for i in range(4):
        vertexes.append(random.randint(1,3))
    return vertexes
```

Функция создания тетраэдра с раскрашенными вершинами:

```python
import math
from sage.plot.plot3d.shapes import Sphere, Text

def Tetrahedron(vertices, color='white'):
    faces = [(0,1,2), (0,1,3), (0,2,3), (1,2,3)]
    return polygons3d(faces, vertices, color=color, opacity = 0.5)

def composition(vertexes):
    colorDict = {1 : 'red', 2 : 'green', 3 : 'blue'}
    T1 = Tetrahedron([(0,0,0), (math.sqrt(3)/2, 0.5, 0), (0,1,0), (math.sqrt(3)/6, 0.5,
                                         math.sqrt(2)/math.sqrt(3))], color='yellow')
    S = Sphere(.05, color=colorDict[vertexes[0]], opacity = 0.7).translate(0,0,0)
    S1 = Sphere(.05, color=colorDict[vertexes[1]], opacity = 0.7).translate(0,1,0)
    S2 = Sphere(.05, color=colorDict[vertexes[2]], opacity = 0.7).translate(math.sqrt(3)/6, 0.5,
                                         math.sqrt(2)/math.sqrt(3))
    S3 = Sphere(.05, color=colorDict[vertexes[3]]).translate(math.sqrt(3)/2, 0.5, 0)
    return T1 + S + S1 + S2 + S3
```

## Решение задачи в общем виде.

Пусть всего будет $m$ цветов. Тогда:  
$|X^e| = m^4$  
$|X^{\tau_1}| = m\cdot3 = m^2$  
Аналогично   
$|X^{\tau_1^2}| = |X^{\tau_2}| = |X^{\tau_2^2}| = |X^{\tau_3}| = |X^{\tau_3^2}| = |X^{\tau_4}| = |X^{\tau_4^2}| = m^2$
$|X^{\sigma_1}| = |X^{\sigma_2}| = |X^{\sigma_3}| = m^2$
Тогда количество раскрасок $N$ по лемме Бернсайда:  
$N = \frac{1}{|G|}\sum_{g \in G} |X^{g}| = \frac{1}{12}\cdot(m^4 + 11\cdot m^2)$  
Ответ: $\frac{1}{12}\cdot(m^4 + 11\cdot m^2)$
