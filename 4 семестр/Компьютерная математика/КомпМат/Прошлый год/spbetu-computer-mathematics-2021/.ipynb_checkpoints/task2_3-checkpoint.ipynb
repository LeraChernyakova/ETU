## Основные теоретические положения

#### Def 1.1
Пусть $G$ - группа с нейтральным элементом $e$, а $X$ - множество. $G$ действует на $X$, если задана операция $G \times X \to X$, обладающая для $\forall x \in X$ и $\forall g,h \in G$ следующими свойствами: 

1) $ex = x$

2) $g(hx) = (gh)x$

где $gx$ - это образ пары $(g,x)$.

#####  Def 1.2
Пусть $X$ - множество. Множество всех биективных функций с $X \to X$ с операцией композиции называется симметрической группой на множестве $X$ и обозначается как $S_{X}$

##### Def 1.3
Орбитой элемента $x \in X$ под действием $G$ называется множество $Gx = \{gx | g \in G\}$. Количество элементов в данной орбите называется длиной орбиты. Любые две орбиты либо не пересекаются, либо совпадают. Таким образом, множество $X$ разбивается в дизъюнктивное объединение орбит.

##### Def 1.4
Неподвижными точками элемента $g \in G$ называются те $x \in X$, для которых $gx = x$. Множество неподвижных точек элемента g обозначается через $X^{g}$.

##### Th 1 (Burnside)
Количество орбит действия группы $G$ на множестве $X$ равно:

$\frac{1}{|G|}\sum_{g \in G} |X^{g}|$

##### Lemma 1
Предположим, что $\sigma$ является циклом длины $k$, а $НОД(k,n) = d$. Тогда перестановка $\sigma^{n}$ является произведением d циклов длины $\frac{k}{d}$.

##### Def 1.5
Пусть $I$ - произвольное множество, а $C$ - множество цветов. Раскраской множества $I$ называется функция из $I$ в $C$. Множество всех раскрасок обозначается через $C^{I}$. 

##### Lemma 2
Пусть $G$ группа, действующая на множестве $I$. Тогда формула $gf(i) = f(g^{-1}i)$, где $g \in G$, $i \in I$, а $f \in C^{I}$ задает действие $G$ на $C^{I}$.

##### Lemma 3
Для элемента $g \in G$ и функции $f \in C^{I}$ следующие условия эквивалентны:

1)$gf = f$

2)$f(i) = f(g^{n}i)$, $\forall i \in I $ и $n \in \mathbb{Z}$

3)Если $i,j \in I$ лежат в одном цикле в циклической записи перестановки $\varphi(g)$, то $f(i) = f(j)$ (в терминологии задачи о бусах, это означает, что все места из одного цикла покрашены в одинаковый цвет)

##### Corollary 1
Количество раскрасок из $C^{I}$, которые сохраняет данный элемент $g \in G$ равно $|C^{c(\varphi(g))}|$, где $c(\varphi(g))$ - количество независимых циклов перестановки $\varphi(g)$ (считая циклы длины 1).

##### Задание
Сколькими способами можно составить бусы из 77 бусинок 4 различных цветов (с точностью до поворотов и осевых симметрий)?

##### Решение
Итак, имеем упорядоченный набор $\{ x_{1}, x_{2}, x_{3}, ... , x_{77} \}$, где $x_{i} \in \{ a, b, c, d \}$. Т.е. 77 бусинок, каждая из которых может быть раскрашена в 4 различных цвета. Существует всего $4^{77}$ таких раскрасок, но не все из них соответствуют различным бусам.

По условию есть две операции, при применении которых будут получаться одинаковые бусы, при повороте и симметричного отображения. Допустим имеем на первом месте красный шар, а на всех остальных белые, применяем поворот(на 60 градусов) получаем на втором месте красный шар, а на всех остальных белые. Бусы до применения операции такие же как и после. 

Допустим имеем бусы $\{1, 2, 3, 4, 5, 6\}$ симметричное отображения будет менять местами бусинки $1-6, 2-5, 3-4$, тогда получится буса $\{6, 5, 4, 3, 2, 1\}$. Бусы до применения операции и после считаем одинаковыми.

Таким образом на множестве всех раскрасок $X$ действует группа поворотов и осевых симметрий. Операция в группе - композиция, она является а ссоциативной, нейтральный элемент - тождественное отображение, обратный элемент к осевой симметрии - сама симметрия (отобразили симметрично $\{1, 2, 3, 4\}$, получили $\{4, 3, 2, 1\}$ затем еще раз и получили исходную перестановку $\{1, 2, 3, 4\}$), обратный элемент к повороту - поворот в другую сторону. Эту группу обозначим как $G$.

Группа $G$ действует на множестве мест $\{1, 2, 3, ..., 77\}$, задано отображение из $G$ в симметрическую группу $S_{77}$. Через $\tau$ обозначим поворот, переводящий каждое место в следующее(двигаем вправо), а последний элемент передвигаем на первый. Через $\sigma$ обозначим осевую симметрию относительно оси проходящую между 1 и 77, и 39.

Для решения нашей задачи, нам необходимо знать сколько циклов в каждой перестановке, который получаются по действием элементов группы $G$. Если бы у нас было количество элементов в перестановке не $77$, а $5$ то может быть нагляднее и практичней было бы просто найти все такие перестановки. Но идея расписывать более $150$ перестановок уже не такая хорошая. К счастью, существует Лемма 1, которая поможет нам найти количество циклов перестановок, получающиеся с помощью применения операции $\tau$, для операций типа $\sigma \tau^{k}$ будет использована другая идея.

Для перестановки $\{ 1, 2, 3, ... ,77 \}$ после применения поворота(операции $\tau$) получается перестановка $\{ 77, 1, 2, 3, ..., 76 \}$. Цикл соответствующий этой перестановке будет $(1,2,3,4,...,77)$ и его длина очевдино равна $k = 77$. Теперь для нахождения количества циклов в перестановках $\tau^{n}$ будет применяться формула $НОД(k,n) = d$. 

Заметим, что множители числа $77$ это простые числа $7$ и $11$. Т.е. $НОД$ отличный от единицы будет получаться при $n \in \{ 7, 14, 21, 28, 35, 42, 49, 56, 63, 70, 11, 22, 33, 44, 55, 66\}$ Тогда $d$ будет принимать значения: $D = (d_{7},d_{14},...,d_{70},d_{11},...,d_{66})$ $D = (7,7,7,7,7,7,7,7,7,7,11,11,11,11,11,11)$

Теперь рассмотрим перестановку после операции $\sigma$. Мы как-бы переводим последний элемент в первый, предпоследний во второй, ... а элемент по середине оставляем на месте. Т.е. этой операции будет соответствовать перестановка $(1, 77)(2, 76)(3, 75)...(39)$, количество циклов равно 39. Надо понять, почему после применении к этой перестановке операций вида $\tau^{k}$ не изменят количество циклов. После такого симметричного отображения для каждого последующего поворота у нас всегда найдется элемент, который на следующем повороте "встанет" на свое место. Т.е. в самом первом повороте встанет на место число 1, потому что он сейчас в самом конце, в последующем двойка и.т.д. Также из-за самой природы симметрии, из-за такой структуры, у нас все остальные циклы будут иметь длину равной двум.

Раскраски, совмещаемые друг с другом каким-нибудь преобразованием из группы $G$ задают одинаковые бусы. Тогда можно сказать, что раскраски задают одинаковые бусы тогда и только тогда, когда они лежат в одной орбите под действием $G$. Таким образом, наша задача состоит в том, чтобы вычислить количество орбит действия $G$ на $X$.

Для того, чтобы воспользоваться леммой Бёрнсайда, необходимо выяснить, сколько элементов множества $X$ оставляет на месте каждый элемент группы $G$. И в этом нам поможет Следствие 1, которое говорит о том, что для подсчёта таких элементов множества $X$ нужно лишь знать количество независимых циклов. Мы как раз уже их подсчитали:

$|X^{e}| = 4^{77}$

$|X^{\tau^{i}}| = 4^{1}$, где $i \in \{1..76\} / \{ i | (i\%7 = 0) || (i\%11 = 0)\}$

$|X^{\tau^{7}}| = |X^{\tau^{14}}| = |X^{\tau^{21}}| = ... = |X^{\tau^{70}}| = 4^{7}$

$|X^{\tau^{11}}| = |X^{\tau^{22}}| = ... = |X^{\tau^{66}}| = 4^{11}$

$|X^{\sigma}| = |X^{\sigma \tau}| = |X^{\sigma \tau^{2}}| = ... = |X^{\sigma \tau^{76}}| = 4^{39}$

$|G| = 154$


Наконец, по лемме Бёрнсайда количество различных бус (число орбит) равно:

$\frac{1}{154}(4^{77} + 60*4^{1} + 10*4^{7} + 6*4^{11} + 77*4^{39}) = 148285474566852974655557450215711063595859544$

### Реализация функции отображения раскраски по заданному номеру

```python
def rotate(perm):
    return [perm[-1]] + perm[:-1]

def symmetry(perm):
    return perm[::-1]
```

```python
def find_perm(perm, index):
    for idx in range(0, index):
        r = random.randint(1,2)
        if r == 1:
            perm = rotate(perm)
            print("rotation")
        else:
            perm = symmetry(perm)
            print("symmetry")
    return perm
```

### Решение задачи в общем виде

Т.е. у нас есть упорядоченный набор $\{x_{1},x_{2}, ..., x_{n}\}$, где $x_{i} \in \{ 1,2,...,m \}$

Затем мы считаем:

$|X^{e}| = m^{n}$

$|X^{\tau^{i}}| = m^{НОД(i,n)}$, где $i \in \{1..n-1\}$

Если $n$ - нечётное число, то:

$|X^{\sigma}| = |X^{\sigma \tau}| = |X^{\sigma \tau^{2}}| = ... = |X^{\sigma \tau^{n-1}}| = m^{\lceil \frac{n}{2} \rceil}$

Если $n$ - чётное, то:

$|X^{\sigma}| = |X^{\sigma \tau^{2}}| = |X^{\sigma \tau^{4}}| = ... = |X^{\sigma \tau^{n-2}}| = m^{\frac{n}{2}}$

и

$|X^{\sigma \tau}| = |X^{\sigma \tau^{3}}| = |X^{\sigma \tau^{5}}| = ... = |X^{\sigma \tau^{n-1}}| = m^{\frac{n}{2}+1}$

Происходит такое разделение на чётное/нечётное $n$ из-за того, что в чётном случае через раз остаются по два цикла единичной длины и из-за этого кол-во циклов возрастает на 1.


После подсчёта этих значений, у нас есть вся необходимая информация для того, чтобы воспользоваться леммой Бёрнсайда:

$\frac{1}{2n}(m^{n} + m^{НОД(1,n)} + m^{НОД(2,n)} + ... + m^{НОД(n-1,n)} + ((n+1)\%2)*(n*m^{\lceil \frac{n}{2} \rceil}) + (n\%2)*(\frac{n}{2}*m^{\frac{n}{2}} + \frac{n}{2}*m^{\frac{n}{2} + 1}))$
