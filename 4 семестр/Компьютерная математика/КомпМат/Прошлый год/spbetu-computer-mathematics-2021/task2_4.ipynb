#### Лемма Бёрнсайда

Количество орбит действия группы $\mathbb G$ на множестве $\mathbb X$ равно $$N = \frac{1}{\# \mathbb G} * \sum_{ g \in \mathbb G} \# \mathbb X^g$$

Множество всех раскрасок: $\mathbb C ^{ I}$.

Количество раскрасок $= (\# \mathbb C)^{\#  I}$

#### Определение 6

Пусть $\mathbb G $ - группа, действующая на множестве $I$ . Тогда формула 

$$gf(i)=f(g^{-1}i),$$ 

где $g \in \mathbb G$, $i \in  I$, а $f \in \mathbb C ^{ I},$  задает действие $g$ на $\mathbb C^{ I}$. 

#### Теорема 

Для элемента $g \in \mathbb G$ и функции $f \in \mathbb C ^I$ следующие условия эквивалентны:

1) $gf=f$

2) $f(i) = f(g^n i) ,\forall i \in I , n \in \mathbb Z$

3) Если $i,j \in I$  лежат в одном цикле в циклической записи перестановки $\varphi (x)$, то  $f(i) = f(j)$. (В терминологии задачи о бусах это означает, что все места из одного цикла покрашены в одинаковый цвет).

#### Определение 7

#### Задание варианта

Сколькими способами можно составить бусы из 161 бусинок 6 различных цветов ( с точностью до поворотов и осевых симметрий)?

#### Решение

Дано множество цветов $\{a,b,c,d,e,f\}$ и множество бусинок {$x_1$,$x_2$,...,$x_{161}$} $\forall i\  x_i \in \{a,b,c,d,e,f\}$. Всего количество бус $6^{161}$. Но необходимо найти количество бус, которые можно составить из этих бусинок, при том, что бусы полученные в результате поворота или осевой симметрии из исходной являются теми же самыми бусами. 

На множестве $X$ всех раскрасок действует группа поворотов и осевых симметрий. Операцией в этой группе является композиция. Операция является ассоциативной, нейтральный элемент - тождественное отображение, обратным к осевой симметрии является она сама, обратным к повроту - поворот в обратную сторону. Обозначим группу буквой $G$.

Задано отображение из $G$ в симметрическую группу $S_{161}$. Пусть $\tau$ - поворот, переводящий каждое место в следующее. $\sigma$ - осевая симметрия.

Раскраски, совмещаемые друг с другом каким-нибудь преобразованием из группы $G$ задают одинаковые бусы. Т.е. раскраски задают одинаковые бусы, когда они лежат в одной орбите под действием $G$. Таким образом задача состоит в том, чтобы вычислить количество орбит действия $G$ на $X$.

Выпишем некоторые перестановки из $S_{161}$ в которые переходят элементы группы $G$:

$e =  \begin{pmatrix}
  1 & 2 & 3 & ... & 161\\
  1 & 2 & 3 & ... & 161
 \end{pmatrix} = (1)(2)(3)...(161)$ 
 
$\tau = \begin{pmatrix}
  1 & 2 & 3 & ... & 161\\
  2 & 3 & 4 & ... & 1
 \end{pmatrix} = (1,2,3,...,161)$
 
 $\tau^2 = \begin{pmatrix}
  1 & 2 & 3 & ... & 160 & 161\\
  3 & 4 & 5 & ... & 1   & 2
 \end{pmatrix} = (1,3,5,...,159,161,2,4,...,160)$
 
 $\sigma = \begin{pmatrix}
  1 & 2 & 3 & ... & 161\\
  161 & 160 & 159 & ... & 1
 \end{pmatrix} = (1,161)(2,160)(3,159)...(79,83)(80,82)(81)$
 

$\sigma^2 = e$ 


Перестановка $\tau$ является циклом длины 161, исходя из Леммы 2 можно посчитать количество циклов в перестановке $\tau^i\ \forall i$. Число $161$ раскладывается на простые множители как $161 = 7*23$. Каждая перестановка $\tau^i$ является произведением $d$ независимых циклов, где $d:= НОД(161,i)$. Следовательно для всех $i$ не кратных 7 или 23 $d$ будет равняться 1. 

Перестановка $\sigma$ является произведением 81 независимого цикла. Рассмотрим из скольки циклов состоят перестановки вида $\sigma \tau^i$.

$\tau^i$ переводит элемент на месте $k$ на место $((k+i-1)\ mod\ 161 + 1)$. 

$\sigma $ переводит элемент $k$ в $(162-k)$. 

$\sigma \tau^i$ переводит $k$ в $(161 - (k+i-1)\ mod\ 161)$ 

$\sigma \tau^i$ переводит $(161-(k+i-1)\ mod \ 161)$ в $k$.

Т.е. перестановки $\sigma \tau^i$ либо переводит элемент сам в себя, либо переводит 2 элемента друг в друга. 

Чтобы понять сколько элементов переходят сами в себя можно составить уравнение $k=161 - (k+i-1) \ mod \ 161$. Это линейное уравнение с одним корнем 

$k = (80-\frac{i}{2}) \ mod \ 161 + 1$.  

$(\frac{1}{2} \ mod \ 161 = 81)$.

Следовательно только один элемент перейдет сам в себя, а остальные 160 элементов будут образовывать циклы длины 2 с какими-нибудь другими элементами. Значит количество независимых циклов в перестановке $\sigma \tau^i$ = 81.


Например, 

$\sigma \tau = \begin{pmatrix}
  1 & 2 & 3 & ... &158& 159 & 160 & 161\\
  160 & 159 & 158 & ...&3& 2  &  1   & 161
 \end{pmatrix} = (1,160)(2,159)...(161)$
 
 
Теперь из определения 7 мы можем найти количество раскрасок, которые сохраняет данный элемент $g \in G$:

$\#X^e = 6^{161}$

$\# X^{\tau^k} = 6$ , $\forall k \in \{1,...,160\}$ : $k \% 7 \neq 0\  \&\& \ k \% 23 \neq 0$

$\# X^{\tau^l} = 6^7$, $\forall l \in \{7,14,21,28,...,154\}$

$\# X^{\tau^m} = 6^{23}$, $\forall m \in \{23,46,69,92,115,138\}$

$\# X^{\sigma} = \# X^{\sigma \tau} = \# X^{\sigma \tau^2} = ... = \# X^{\sigma \tau^{160}} = 6^{81}$ 

$\# G = 322$


Теперь мы можем воспользоваться леммой Бернсайда: количество различных бус равно $$\frac{1}{322}(6^{161} + 6*132 +6^{7}*22 + 6^{23}*6 + 6^{81}*161)$$
