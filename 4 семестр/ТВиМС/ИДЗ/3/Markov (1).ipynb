```python
def get_seq(steps):
    start = 1
    P = [[0.4,0.1,0,0.2,0.3],
         [0.2,0.1,0.4,0.3,0],
         [0,0.4,0,0.4,0.2],
         [0.6,0.1,0.1,0.2,0],
         [0.2,0.3,0.3,0,0.2]]
    E = [[0 if x!= i else 1 for x in range(5)]for i in range(5)]
    a = hmm.DiscreteHiddenMarkovModel(P,E,[1,2,3,4,5])
    seq = [*(a.generate_sequence(steps,starting_state=start)[0])]
    return seq
```
