### Задание 6.

```python
def get_seq(steps, start_point):
    start = start_point
    P = [[0.1, 0.2, 0.3, 0, 0.4, 0],
         [0.4, 0.1, 0.2, 0, 0.3, 0],
         [0.4, 0.1, 0.3, 0, 0.2, 0],
         [0, 0.2, 0, 0.4, 0.4, 0],
         [0.1, 0.2, 0.5, 0, 0.2, 0],
         [0, 0, 0, 0, 0.8, 0.2]]
    E = [[0 if x!= i else 1 for x in range(6)] for i in range(6)]
    a = hmm.DiscreteHiddenMarkovModel(P,E,[1, 2, 3, 4, 5, 6])
    seq = [*(a.generate_sequence(steps,starting_state=start)[0])]
    return seq
```

```python
print('длина = 10, старт = 1')
print(','.join(list(map(str,get_seq(10, 0)))))
```

```python
start_point = 0
for steps in [10,50,100,1000]:
    seq = get_seq(steps, start_point)
    x_nums = [0 for i in range(6)]
    for elem in seq:
        x_nums[elem] += 1
    x_nums = list(map(lambda t: t/steps,x_nums))
    print(f'Моделирование цепи маркова длинны {steps}')
    print('Старт ', start_point+1)
    print(','.join(list(map(lambda t: str(t + 1),seq))))
    print(x_nums)
    print('\n')
    start_point += 1
```
