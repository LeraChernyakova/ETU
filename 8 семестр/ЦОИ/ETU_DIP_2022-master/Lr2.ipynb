```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
loading_filename = 'picture.jpg'
result_filename = 'result.bmp'
```

```python
def show_image(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.show()
```

```python
image = cv2.imread(loading_filename)
show_image(image)
```

```python
min_dis = 100
min_rad = 0
max_rad = 0


gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
show_image(gray)
gray = cv2.GaussianBlur(gray,(5,5),cv2.BORDER_DEFAULT)
show_image(gray)

circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, 1.5, minDist=min_dis,
                           minRadius=min_rad, maxRadius=max_rad)
if circles is not None:
    circles = np.round(circles[0, :]).astype("int")
    for (x, y, r) in circles:
        cv2.circle(image, (x, y), r, (0, 255, 0), 4)
        cv2.rectangle(image, (x - 5, y - 5), (x + 5, y + 5), (0, 128, 255), -1)
show_image(image)
```

```python
cv2.imwrite(result_filename, image)
```
