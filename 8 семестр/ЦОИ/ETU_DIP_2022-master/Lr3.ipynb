```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
loading_filename = 'picture.jpg'
result_filename = 'result.bmp'
```

```python
def show_image(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.show()
```

```python
image = cv2.imread(loading_filename)
show_image(image)
```

```python
k = 5
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

pixel_values = image.reshape((-1, 3))
pixel_values = np.float32(pixel_values)

criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
ret,label,center = cv2.kmeans(pixel_values, k, None, criteria, 10, cv2.KMEANS_PP_CENTERS)

center = np.uint8(center)
res = center[label.flatten()]
result_image = res.reshape((image.shape))
show_image(result_image)
```

```python
cv2.imwrite(result_filename, result_image)
```
