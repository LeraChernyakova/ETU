Реализовать на языке Python  с использованием библиотеки OpenCV программу, выполняющее сегментацию/кластеризацию изображения по цвету с использованием метода  K-Means

```python
#  Основные импорты
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

#### Инициализация основных переменных

```python
img_in = 'input.jpg'
k = 5
img_out = 'result.bmp'
```

#### Открытие изображения

```python
img = cv2.imread(img_in)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

#### Сегментация изображения методом K-Means

```python
# Преобразование изображения: массив пикселей с RGB значениями
pixels = img.reshape((-1,3))
pixels = np.float32(pixels)

# Критерий остановки
stop = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.85) 

# Кластеризация
_, pixels_cluster, centers_cluster = cv2.kmeans(pixels, k, None, stop, 10, cv2.KMEANS_RANDOM_CENTERS) 

# Замена пикселей на цвета центров кластеров
segmented_img = np.uint8(centers_cluster)[pixels_cluster.flatten()].reshape((img.shape)) 

plt.imshow(segmented_img)
plt.show()
cv2.imwrite(img_out, cv2.cvtColor(segmented_img, cv2.COLOR_BGR2RGB))
```
