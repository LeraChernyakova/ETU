### Задача
Реализовать на языке Python  с использованием библиотеки OpenCV программу, выполняющее поиск окружностей в изображении путем преобразования Хо (Hough)

```python
#  Основные импорты
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

#### Инициализация основных переменных

```python
img_in = 'input.jpg'
minD = 200
minR = 40
maxR = 125
img_out = 'result.bmp'
```

#### Открытие изображения

```python
img = cv2.imread(img_in)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

#### Поиск окружностей

```python
# Преобразование в оттенки серого + уменьшение шума
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gray = cv2.medianBlur(img_gray, 5)

# Поиск кругов путем преобразования Хо
circles = cv2.HoughCircles(img_gray, cv2.HOUGH_GRADIENT, dp = 1.4, param1 = 80, param2 = 57, minDist = minD, minRadius = minR, maxRadius = maxR)

# Отрисовка окружностей
if circles is not None:
    circles = np.round(circles[0, :]).astype(int)
    for i in circles:
        center = (i[0], i[1])
        radius = i[2]
        cv2.circle(img, center, 1, (255, 0, 0), 10)  
        cv2.circle(img, center, radius, (255, 0, 0), 10)

plt.imshow(img)
plt.show()

cv2.imwrite(img_out, cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
